## Capítulo 1: Estatística Descritiva

Estatística descritiva resume um conjunto de dados em poucos números. As **medidas
de tendência central** dizem "onde os dados se concentram": média (`mean`), mediana (`median`, o
valor do meio) e moda (`mode`, o valor mais frequente). As **medidas de dispersão** dizem "o quanto
os dados variam": desvio padrão (`std`), variância (`var`) e os quartis (que dividem os dados em 4
partes iguais — o intervalo entre o 1º e 3º quartil é chamado de IQR).

📌 A mediana é mais **robusta** a valores extremos (outliers) do que a média — por isso salários e
preços de imóveis costumam ser resumidos pela mediana.

**💻 Exemplo comentado:**

````python
import numpy as np
import pandas as pd

notas = pd.Series([7, 8, 6, 9, 7, 5, 8, 10, 7, 100])  # 100 é um outlier proposital

print(f"Média: {notas.mean():.2f}")
print(f"Mediana: {notas.median():.2f}")     # bem menos afetada pelo outlier
print(f"Moda: {notas.mode().tolist()}")
print(f"Desvio padrão: {notas.std():.2f}")
print(f"Quartis: {notas.quantile([0.25, 0.5, 0.75]).tolist()}")
````

## Capítulo 2: Distribuições de Probabilidade

Uma distribuição de probabilidade descreve como os valores de uma variável se
espalham. A mais famosa é a **distribuição normal** (o "sino"), usada para descrever alturas, erros
de medição, etc. A **distribuição binomial** descreve o número de sucessos em `n` tentativas
independentes (ex.: número de caras em 10 lançamentos de moeda). A biblioteca `scipy.stats` calcula
probabilidades dessas distribuições sem fórmulas manuais.

**💻 Exemplo comentado:**

````python
from scipy import stats
import numpy as np

# Distribuição normal com média 170 e desvio padrão 10 (ex.: alturas em cm)
prob_menor_180 = stats.norm.cdf(180, loc=170, scale=10)
print(f"P(altura < 180cm): {prob_menor_180:.3f}")

# Distribuição binomial: 10 lançamentos de moeda, probabilidade de sucesso 0.5
prob_exatos_5 = stats.binom.pmf(5, n=10, p=0.5)
print(f"P(exatamente 5 caras em 10 lançamentos): {prob_exatos_5:.3f}")

# Gerando uma amostra aleatória normal
np.random.seed(42)
amostra = np.random.normal(loc=170, scale=10, size=1000)
print(f"Média da amostra gerada: {amostra.mean():.2f}")
````

## Capítulo 3: Testes de Hipóteses

Um teste de hipótese responde: "essa diferença que eu observei é real, ou pode ser
só acaso?". Definimos uma **hipótese nula** (H0: "não há diferença") e calculamos o **valor-p** — a
probabilidade de observar um resultado tão extremo quanto o nosso, se H0 fosse verdadeira. Por
convenção, se valor-p < 0.05, rejeitamos H0 (a diferença é "estatisticamente significativa").

📌 O teste t (`scipy.stats.ttest_ind` para duas amostras independentes, `ttest_1samp` para comparar
uma amostra a um valor fixo) é um dos testes mais usados para comparar médias.

**💻 Exemplo comentado:**

````python
from scipy import stats
import numpy as np

np.random.seed(0)
grupo_a = np.random.normal(loc=50, scale=5, size=30)   # ex.: notas da turma A
grupo_b = np.random.normal(loc=53, scale=5, size=30)   # ex.: notas da turma B

estatistica, valor_p = stats.ttest_ind(grupo_a, grupo_b)
print(f"Estatística t: {estatistica:.3f}")
print(f"Valor-p: {valor_p:.3f}")

if valor_p < 0.05:
    print("Diferença estatisticamente significativa (rejeitamos H0)")
else:
    print("Não há evidência suficiente de diferença (não rejeitamos H0)")
````

## Capítulo 4: Correlação e Covariância

**Covariância** mede se duas variáveis crescem juntas (positiva), uma cresce
enquanto a outra diminui (negativa), ou não têm relação clara (perto de zero) — mas sua escala
depende das unidades, o que dificulta comparação. A **correlação de Pearson** resolve isso: é sempre
um número entre -1 e 1, fácil de interpretar (perto de 1: forte relação positiva; perto de -1: forte
relação negativa; perto de 0: sem relação linear).

⚠️ **Correlação não implica causalidade!** Duas variáveis podem estar correlacionadas por
coincidência ou por uma terceira variável que afeta as duas — sempre desconfie antes de concluir que
uma causa a outra.

**💻 Exemplo comentado:**

````python
import pandas as pd

df = pd.DataFrame({
    "horas_estudo": [1, 2, 3, 4, 5, 6],
    "nota": [50, 55, 65, 70, 80, 90],
})

print(df.corr())              # matriz de correlação entre todas as colunas numéricas
print(df["horas_estudo"].corr(df["nota"]))   # correlação entre duas colunas específicas
print(df.cov())                # matriz de covariância
````

## Capítulo 5: Amostragem e Intervalos de Confiança

Raramente conseguimos analisar uma população inteira — em vez disso, coletamos uma
**amostra** (`df.sample()` retira linhas aleatórias) e usamos ela para estimar características da
população. Um **intervalo de confiança (IC)** expressa essa incerteza: em vez de dizer "a média é
exatamente X", dizemos "temos 95% de confiança de que a média populacional está entre A e B" — mais
honesto do que um número único.

**💻 Exemplo comentado:**

````python
import numpy as np
from scipy import stats

np.random.seed(1)
amostra = np.random.normal(loc=50, scale=10, size=40)   # amostra de 40 observações

media = np.mean(amostra)
erro_padrao = stats.sem(amostra)                          # erro padrão da média
intervalo = stats.t.interval(0.95, df=len(amostra) - 1, loc=media, scale=erro_padrao)

print(f"Média da amostra: {media:.2f}")
print(f"Intervalo de confiança de 95%: ({intervalo[0]:.2f}, {intervalo[1]:.2f})")
````

### ✅ Gabarito do projeto

````python
import numpy as np
from scipy import stats

np.random.seed(3)
campanha_a = np.random.normal(loc=40, scale=8, size=25)
campanha_b = np.random.normal(loc=48, scale=8, size=25)
gasto_ads = np.random.uniform(100, 500, 25)
conversoes = 20 + gasto_ads * 0.08 + np.random.normal(0, 3, 25)

# 1. Estatística descritiva
print("Campanha A - média:", round(campanha_a.mean(), 2), "| mediana:", round(np.median(campanha_a), 2), "| desvio:", round(campanha_a.std(), 2))
print("Campanha B - média:", round(campanha_b.mean(), 2), "| mediana:", round(np.median(campanha_b), 2), "| desvio:", round(campanha_b.std(), 2))

# 2. Teste de hipótese
estatistica, valor_p = stats.ttest_ind(campanha_a, campanha_b)
print(f"\nValor-p do teste t: {valor_p:.4f}")
if valor_p < 0.05:
    vencedora = "B" if campanha_b.mean() > campanha_a.mean() else "A"
    print(f"Diferença estatisticamente significativa! Campanha vencedora: {vencedora}")
else:
    print("Não há diferença estatisticamente significativa entre as campanhas.")

# 3. Correlação entre gasto e conversões
correlacao = np.corrcoef(gasto_ads, conversoes)[0, 1]
print(f"\nCorrelação entre gasto em ads e conversões: {correlacao:.3f}")

# 4. Intervalo de confiança da campanha vencedora (B, neste caso)
media_b = campanha_b.mean()
erro_b = stats.sem(campanha_b)
intervalo_b = stats.t.interval(0.95, df=len(campanha_b) - 1, loc=media_b, scale=erro_b)
print(f"IC 95% da campanha B: ({intervalo_b[0]:.2f}, {intervalo_b[1]:.2f})")
````
