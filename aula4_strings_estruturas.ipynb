## 📘 Teoria dialogada (30 min)

**— Aluno(a):** Dados do mundo real vêm sempre arrumadinhos, certo?

**— Professor(a):** Quase nunca! Nomes com espaços extras, maiúsculas misturadas, textos com erros de
digitação... por isso strings têm tantos métodos prontos: `.strip()` remove espaços das pontas,
`.replace()` troca um trecho por outro, `.upper()`/`.lower()` padronizam maiúsculas/minúsculas,
`.split()` separa em uma lista de palavras e `.count()` conta ocorrências. E dá para "fatiar" uma
string com `[inicio:fim]`, ou até inverter com `[::-1]`.

**— Aluno(a):** E para guardar vários dados juntos?

**— Professor(a):** Aí entram as três estruturas fundamentais. Uma **lista** (`[]`) é editável e
ordenada — como uma lista de compras: você usa `.append()` para adicionar, `.remove()` para tirar,
`.sort()` para ordenar, `.pop()` para remover pelo índice e `.index()` para descobrir a posição de um
item. Uma **tupla** (`()`) é parecida, mas **imutável** — uma vez criada, não muda mais (ótima para
dados que não devem ser alterados, como coordenadas). E um **dicionário** (`{}`) guarda pares
chave-valor — como uma agenda, onde o nome leva ao telefone. Para percorrer um dicionário, usamos
`.items()` (chave e valor juntos), `.keys()` (só as chaves) ou `.values()` (só os valores).

**— Aluno(a):** E se eu tiver uma lista inteira de "fichas", tipo vários produtos com nome e preço?

**— Professor(a):** Aí você usa uma **lista de dicionários** — é assim que representamos dados
tabulares (como uma planilha) em Python puro, e é exatamente o que vamos praticar hoje com uma lista
de vendas.


**💻 Exemplo comentado:**

````python
frase = "  Python é incrível "
print(frase.strip().upper())              # remove espaços e deixa maiúsculo
print(frase.strip().replace("incrível", "poderoso"))

frutas = ["maçã", "banana"]
frutas.append("uva")                       # adiciona ao final
frutas.sort()                              # ordena
print(frutas)

coordenadas = (10.5, -23.4)                # tupla: imutável

contato = {"nome": "Ana", "telefone": "9999-0000"}
for chave, valor in contato.items():
    print(chave, "->", valor)
````

## 🧑‍💻 Prática guiada (30 min)

### 🧹 Limpeza de nomes com espaços

````python
nomes_sujos = ["  ana silva ", "JOÃO   PEDRO", " maria eduarda", "carlos  alberto "]

nomes_limpos = []
for nome in nomes_sujos:
    nome_tratado = nome.strip().replace("  ", " ").title()
    nomes_limpos.append(nome_tratado)

print(nomes_limpos)
````

### 📇 Agenda de contatos

````python
agenda = [
    {"nome": "Ana", "telefone": "9999-0001"},
    {"nome": "Bruno", "telefone": "9999-0002"},
]

def adicionar_contato(agenda, nome, telefone):
    agenda.append({"nome": nome, "telefone": telefone})

def buscar_contato(agenda, nome):
    for contato in agenda:
        if contato["nome"].lower() == nome.lower():
            return contato
    return None

adicionar_contato(agenda, "Carla", "9999-0003")
print(agenda)
print(buscar_contato(agenda, "bruno"))
````

## 📌 Checkpoint — Comandos da Aula 4

| Comando / Conceito | O que faz |
|---|---|
| `.strip() / .replace() / .split()` | Limpeza e transformação de strings |
| `.upper() / .lower() / .title()` | Padronização de maiúsculas/minúsculas |
| `.count()` | Conta ocorrências de um caractere/substring |
| `[inicio:fim] e [::-1]` | Fatiamento e inversão de strings/listas |
| `.append() / .remove() / .pop() / .index()` | Métodos de lista |
| `.sort()` | Ordena uma lista |
| `tupla ()` | Coleção ordenada e imutável |
| `.items() / .keys() / .values()` | Percorrer um dicionário |
| `zip()` | Combina duas listas elemento a elemento |

````python
# --- Exercício 1 ---
nome_completo = "João Pedro Almeida"
print(nome_completo.split()[0])

# --- Exercício 2 ---
palavra = "banana"
print(palavra.count("a"))

# --- Exercício 3 ---
palavra = "python"
print(palavra[::-1])

# --- Exercício 4 ---
# Tuplas são imutáveis. Se precisamos alterar, devemos usar uma lista:
cores = ["azul", "dourado"]
cores[0] = "verde"
print(cores)

# --- Exercício 5 ---
nomes = ["  ana silva ", "JOÃO   PEDRO", " maria eduarda"]
nomes_limpos = [n.strip().replace("  ", " ").title() for n in nomes]
print(nomes_limpos)

# --- Exercício 6 ---
numeros = [30, 10, 20]
numeros.append(5)
numeros.pop(1)
numeros.sort()
print(numeros)

# --- Exercício 7 ---
vendas = [
    {"produto": "Caneta", "qtd": 10, "preco": 2.5},
    {"produto": "Caderno", "qtd": 4, "preco": 15.0},
    {"produto": "Caneta", "qtd": 6, "preco": 2.5},
]
total = sum(v["qtd"] * v["preco"] for v in vendas)
print(f"Total vendido: R$ {total:.2f}")

# --- Exercício 8 ---
livro = {"titulo": "Python Fluente", "autor": "Luciano Ramalho", "ano": 2015}
for chave, valor in livro.items():
    print(chave, ":", valor)

# --- Desafio 1 ---
nomes = ["Ana", "Bruno", "Carla"]
telefones = ["9999-1", "9999-2", "9999-3"]
agenda = dict(zip(nomes, telefones))
print(agenda)

# --- Desafio 2 ---
agenda = [{"nome": "Ana", "telefone": "1"}, {"nome": "Bruno", "telefone": "2"}]

def buscar_contato(agenda, nome):
    for contato in agenda:
        if contato["nome"].lower() == nome.lower():
            return contato
    return None

print(buscar_contato(agenda, "ana"))
print(buscar_contato(agenda, "carla"))
````
