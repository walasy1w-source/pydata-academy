## Capítulo 1: Clusterização com K-Means

**K-Means** é o algoritmo de clusterização mais usado do mercado. Funciona de forma
iterativa: (1) escolhe `k` centros aleatórios (centróides), (2) atribui cada ponto ao centróide mais
próximo, (3) recalcula os centróides como a média dos pontos atribuídos a eles, repetindo até
estabilizar. O atributo `.inertia_` mede o quão compactos são os clusters (soma das distâncias ao
quadrado até o centróide mais próximo) — usado no **método do cotovelo** para escolher o número ideal
de clusters `k`.

**💻 Exemplo comentado:**

````python
from sklearn.cluster import KMeans
import numpy as np

np.random.seed(0)
X = np.vstack([
    np.random.normal([0, 0], 0.5, (20, 2)),
    np.random.normal([10, 10], 0.5, (20, 2)),
    np.random.normal([0, 10], 0.5, (20, 2)),
])

modelo = KMeans(n_clusters=3, random_state=0, n_init=10)
rotulos = modelo.fit_predict(X)

print("Centróides encontrados:\n", modelo.cluster_centers_.round(2))
print("Inércia (compactação dos clusters):", round(modelo.inertia_, 2))

# Método do cotovelo: testando vários valores de k
for k in [1, 2, 3, 4, 5]:
    m = KMeans(n_clusters=k, random_state=0, n_init=10).fit(X)
    print(f"k={k}: inércia={m.inertia_:.2f}")
````

## Capítulo 2: Clusterização Hierárquica

A **clusterização hierárquica** (`AgglomerativeClustering`) constrói uma hierarquia
de clusters: começa tratando cada ponto como seu próprio cluster e vai **unindo** os mais próximos
sucessivamente, formando uma árvore (dendrograma). Diferente do K-Means, não exige escolher `k` de
antemão (embora o scikit-learn peça um `n_clusters` para "cortar" a árvore em algum ponto) e não
depende de inicialização aleatória — o resultado é determinístico.

**💻 Exemplo comentado:**

````python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

np.random.seed(0)
X = np.vstack([
    np.random.normal([0, 0], 0.5, (10, 2)),
    np.random.normal([10, 10], 0.5, (10, 2)),
])

modelo = AgglomerativeClustering(n_clusters=2)
rotulos = modelo.fit_predict(X)
print("Rótulos (clusterização hierárquica):", rotulos)

# Comparando com KMeans no mesmo dataset
from sklearn.cluster import KMeans
modelo_kmeans = KMeans(n_clusters=2, random_state=0, n_init=10)
rotulos_kmeans = modelo_kmeans.fit_predict(X)
print("Rótulos (KMeans):", rotulos_kmeans)
````

## Capítulo 3: Redução de Dimensionalidade (PCA)

**PCA (Principal Component Analysis)** — já visto conceitualmente em Álgebra Linear
(Sprint 11) via autovalores/autovetores — reduz o número de features de um dataset, encontrando as
direções (**componentes principais**) que capturam a maior parte da variância dos dados. Útil para
visualizar dados de muitas dimensões em 2D/3D, acelerar treinamento, e remover redundância entre
features correlacionadas antes de clusterizar. `PCA(n_components=k).fit_transform(X)` retorna os dados
já projetados nas `k` direções principais; `.explained_variance_ratio_` mostra quanta variância cada
componente captura.

**💻 Exemplo comentado:**

````python
from sklearn.decomposition import PCA
import numpy as np

np.random.seed(0)
# Dados com 4 features, mas 2 delas são combinações lineares das outras (redundantes)
X = np.random.rand(100, 2)
X = np.column_stack([X, X[:, 0] * 2, X[:, 1] * 3])   # 4 colunas, mas só 2 "graus de liberdade" reais

pca = PCA(n_components=2)
X_reduzido = pca.fit_transform(X)

print("Shape original:", X.shape, "-> Shape reduzido:", X_reduzido.shape)
print("Variância explicada por componente:", pca.explained_variance_ratio_.round(4))
print("Variância total explicada por 2 componentes:", pca.explained_variance_ratio_.sum().round(4))
````

````python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# 1. Dataset sintético de clientes (sem rótulos)
np.random.seed(42)
n = 150
idade = np.random.randint(18, 70, n)
renda = idade * 150 + np.random.normal(0, 500, n)
frequencia_compra = np.random.randint(1, 30, n)
valor_gasto = frequencia_compra * 25 + np.random.normal(0, 40, n)

clientes = pd.DataFrame({
    "idade": idade,
    "renda": renda,
    "frequencia_compra": frequencia_compra,
    "valor_gasto": valor_gasto,
})
print(f"1. Dataset gerado com {len(clientes)} clientes.")

# 2. Redução de dimensionalidade com PCA
pca = PCA(n_components=2)
X_reduzido = pca.fit_transform(clientes)
print(f"2. Variância explicada pelos 2 componentes: {pca.explained_variance_ratio_.sum():.4f}")

# 3. K-Means nos dados reduzidos
modelo = KMeans(n_clusters=3, random_state=42, n_init=10)
clientes["cluster"] = modelo.fit_predict(X_reduzido)
clusters_encontrados = sorted(clientes["cluster"].unique())
print(f"3. Clusters encontrados: {clusters_encontrados}")

# 4. Interpretação: médias das features originais por cluster
medias_por_cluster = clientes.groupby("cluster")[["idade", "renda", "frequencia_compra", "valor_gasto"]].mean()
print("\n4. Perfil médio de cada cluster:")
print(medias_por_cluster.round(2))
````
