## Capítulo 1: Introdução aos Métodos Numéricos

Muitos problemas não têm uma **solução analítica exata** (uma fórmula fechada) —
ou ela é complexa demais para calcular diretamente. **Métodos numéricos** resolvem esses problemas de
forma **aproximada**, por meio de cálculos iterativos, aceitando um pequeno erro em troca de uma
resposta prática. Exemplo clássico: calcular a raiz quadrada por aproximações sucessivas (método de
Newton), em vez de uma fórmula fechada.

**💻 Exemplo comentado:**

````python
import math

def raiz_quadrada_newton(numero, palpite_inicial=1.0, iteracoes=10):
    x = palpite_inicial
    for _ in range(iteracoes):
        x = 0.5 * (x + numero / x)   # fórmula iterativa do método de Newton
    return x

resultado_aproximado = raiz_quadrada_newton(16)
resultado_exato = math.sqrt(16)

print(f"Aproximação numérica: {resultado_aproximado:.6f}")
print(f"Valor exato (math.sqrt): {resultado_exato:.6f}")
print(f"Erro: {abs(resultado_aproximado - resultado_exato):.10f}")
````

## Capítulo 2: Interpolação

**Interpolação** estima valores **entre** pontos conhecidos. A forma mais simples é
a interpolação **linear** (`np.interp`): assume que entre dois pontos conhecidos, os valores mudam de
forma reta. Para relações mais complexas, podemos ajustar um **polinômio** aos pontos (`np.polyfit`
encontra os coeficientes, `np.polyval` avalia o polinômio em novos pontos) — útil para preencher
lacunas em séries temporais ou estimar valores entre medições.

**💻 Exemplo comentado:**

````python
import numpy as np

# Pontos conhecidos: dias e temperatura medida
dias_conhecidos = np.array([0, 5, 10, 15, 20])
temperaturas = np.array([20, 24, 22, 28, 25])

# Interpolação linear: qual seria a temperatura no dia 7?
temp_dia_7 = np.interp(7, dias_conhecidos, temperaturas)
print(f"Temperatura estimada no dia 7 (linear): {temp_dia_7:.2f}")

# Ajuste polinomial (grau 2) para capturar uma curva mais suave
coeficientes = np.polyfit(dias_conhecidos, temperaturas, deg=2)
temp_dia_7_poly = np.polyval(coeficientes, 7)
print(f"Temperatura estimada no dia 7 (polinômio grau 2): {temp_dia_7_poly:.2f}")
````

## Capítulo 3: Diferenciação Numérica

A **derivada** mede a taxa de variação de uma função. Quando não temos a fórmula
exata (só pontos de dados) ou a derivada analítica é difícil, aproximamos numericamente com
**diferenças finitas**: `f'(x) ≈ (f(x+h) - f(x-h)) / (2h)` (diferença central), usando um `h` pequeno.
O `np.gradient` calcula essa aproximação automaticamente para um array inteiro de valores.

**💻 Exemplo comentado:**

````python
import numpy as np

def f(x):
    return x ** 2   # derivada exata: f'(x) = 2x

def derivada_numerica(func, x, h=1e-5):
    return (func(x + h) - func(x - h)) / (2 * h)

print(f"Derivada numérica em x=3: {derivada_numerica(f, 3):.4f}")
print(f"Derivada exata em x=3 (2x): {2 * 3}")

# Usando np.gradient em um array de valores de uma função
x_valores = np.linspace(0, 10, 100)
y_valores = x_valores ** 2
derivadas = np.gradient(y_valores, x_valores)
print(f"Derivada aproximada (np.gradient) perto de x=3: {derivadas[30]:.2f}")
````

## Capítulo 4: Integração Numérica

A **integral** de uma função calcula a área sob sua curva — útil, por exemplo, para
calcular probabilidade acumulada (Sprint 4) quando não há fórmula fechada. A **regra do trapézio**
(`np.trapz`) aproxima essa área somando pequenos trapézios entre os pontos conhecidos. Para funções
contínuas conhecidas, `scipy.integrate.quad` oferece uma aproximação numérica ainda mais precisa.

**💻 Exemplo comentado:**

````python
import numpy as np
from scipy import integrate

# Área sob y = x^2 entre 0 e 3, usando pontos discretos (regra do trapézio)
x = np.linspace(0, 3, 100)
y = x ** 2
area_trapezio = np.trapz(y, x)
print(f"Área aproximada (trapézio): {area_trapezio:.4f}")

# Usando scipy.integrate.quad para a função contínua (mais preciso)
area_quad, erro_estimado = integrate.quad(lambda x: x ** 2, 0, 3)
print(f"Área com scipy.integrate.quad: {area_quad:.4f}")

# Valor exato (integral de x^2 de 0 a 3 = x^3/3 = 27/3 = 9)
print("Valor exato: 9.0")
````

## Capítulo 5: Resolução de Equações Não Lineares

Muitas equações (`f(x) = 0`) não têm solução algébrica simples (ex.: envolvendo
exponenciais, ou polinômios de grau alto). Métodos numéricos encontram a **raiz** (o valor de `x` que
zera a função) de forma aproximada e iterativa. `scipy.optimize.brentq(f, a, b)` encontra a raiz de
`f` dentro de um intervalo `[a, b]` onde sabemos que a função muda de sinal (garantindo que existe uma
raiz ali).

**💻 Exemplo comentado:**

````python
from scipy import optimize
import numpy as np

# Encontrando a raiz de f(x) = x^2 - 4 (raízes exatas: x=2 e x=-2)
def f(x):
    return x ** 2 - 4

raiz = optimize.brentq(f, 0, 5)   # busca no intervalo [0, 5], onde sabemos que há uma raiz positiva
print(f"Raiz encontrada: {raiz:.6f}")

# Uma equação sem solução algébrica simples: x = cos(x)
def g(x):
    return x - np.cos(x)

raiz_g = optimize.brentq(g, 0, 1)
print(f"Solução de x = cos(x): {raiz_g:.6f}")
````

## Capítulo 6: Otimização Numérica

**Otimizar** significa encontrar o valor de entrada que minimiza (ou maximiza) uma
função — exatamente o que um modelo de ML faz ao "aprender": minimizar o erro. `scipy.optimize.minimize`
encontra numericamente o ponto de mínimo de uma função, a partir de um palpite inicial (`x0`), sem
precisarmos calcular a derivada manualmente.

**💻 Exemplo comentado:**

````python
from scipy import optimize

# Minimizando f(x) = (x - 3)^2 + 5 -> o mínimo é em x=3, valor mínimo = 5
def f(x):
    return (x - 3) ** 2 + 5

resultado = optimize.minimize(f, x0=0)
print(f"Ponto de mínimo encontrado: x = {resultado.x[0]:.4f}")
print(f"Valor mínimo da função: {resultado.fun:.4f}")
````

## Capítulo 7: Gradiente Descendente

O **gradiente descendente** é o algoritmo de otimização por trás do treinamento de
praticamente todo modelo de ML e rede neural. A ideia: partindo de um ponto qualquer, calculamos a
**derivada** (gradiente) da função de custo naquele ponto — ela aponta na direção de **maior
crescimento**. Damos um passo na direção **oposta** (descendente), de tamanho proporcional à **taxa de
aprendizado** (`learning_rate`), repetindo até o valor parar de melhorar significativamente.

**💻 Exemplo comentado:**

````python
def f(x):
    return (x - 3) ** 2   # mínimo em x=3

def derivada_f(x):
    return 2 * (x - 3)   # derivada analítica de (x-3)^2

def gradiente_descendente(derivada, x_inicial, taxa_aprendizado=0.1, iteracoes=50):
    x = x_inicial
    historico = [x]
    for _ in range(iteracoes):
        gradiente = derivada(x)
        x = x - taxa_aprendizado * gradiente   # passo na direção oposta ao gradiente
        historico.append(x)
    return x, historico

x_final, historico = gradiente_descendente(derivada_f, x_inicial=0)
print(f"Valor final de x: {x_final:.6f} (mínimo real: 3)")
print(f"Valor de f no final: {f(x_final):.6f}")
print(f"Primeiros passos: {[round(v, 3) for v in historico[:5]]}")
````

````python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Dados sintéticos com relação linear conhecida (a=2, b=5) mais ruído
np.random.seed(0)
x = np.linspace(0, 10, 50)
y = 2 * x + 5 + np.random.normal(0, 1, 50)

# 2. Gradiente descendente para regressão linear
def erro_quadratico_medio(a, b, x, y):
    previsoes = a * x + b
    return np.mean((previsoes - y) ** 2)

def gradiente_descendente_regressao(x, y, taxa_aprendizado=0.01, iteracoes=1000):
    a, b = 0.0, 0.0
    n = len(x)
    historico_erro = []
    for _ in range(iteracoes):
        previsoes = a * x + b
        erro = previsoes - y
        gradiente_a = (2 / n) * np.dot(erro, x)
        gradiente_b = (2 / n) * np.sum(erro)
        a = a - taxa_aprendizado * gradiente_a
        b = b - taxa_aprendizado * gradiente_b
        historico_erro.append(erro_quadratico_medio(a, b, x, y))
    return a, b, historico_erro

a_final, b_final, historico_erro = gradiente_descendente_regressao(x, y)
print(f"1 e 2. Gradiente descendente -> a={a_final:.3f}, b={b_final:.3f}")

# 3. Comparando com scikit-learn
modelo = LinearRegression()
modelo.fit(x.reshape(-1, 1), y)
print(f"3. Scikit-learn -> a={modelo.coef_[0]:.3f}, b={modelo.intercept_:.3f}")
print("   Resultados próximos?", abs(a_final - modelo.coef_[0]) < 0.1 and abs(b_final - modelo.intercept_) < 0.5)

# 4. Evolução do erro ao longo das iterações
print(f"\n4. Erro na iteração 0: {historico_erro[0]:.3f}")
print(f"   Erro na iteração 500: {historico_erro[500]:.3f}")
print(f"   Erro na última iteração: {historico_erro[-1]:.3f}")
print("   O erro diminuiu ao longo do treinamento:", historico_erro[0] > historico_erro[-1])
````
