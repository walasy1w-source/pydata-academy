## Capítulo 1: Vetores

Um **vetor** é uma lista ordenada de números — em ML, cada linha de um DataFrame é,
matematicamente, um vetor de features. Operações fundamentais: **soma** (elemento a elemento),
**multiplicação por escalar** (multiplica cada elemento pelo mesmo número), **produto escalar** (`dot
product`: soma dos produtos elemento a elemento, usado para medir similaridade), e **norma** (o
"comprimento" do vetor, `np.linalg.norm`).

**💻 Exemplo comentado:**

````python
import numpy as np

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print("Soma:", v1 + v2)
print("Multiplicação por escalar (2x):", v1 * 2)
print("Produto escalar:", np.dot(v1, v2))
print("Norma de v1:", np.linalg.norm(v1))

# Vetor unitário (mesma direção, comprimento 1) — muito usado para comparar direções
v1_unitario = v1 / np.linalg.norm(v1)
print("Vetor unitário de v1:", v1_unitario)
````

## Capítulo 2: Matrizes e Operações

Uma **matriz** é uma tabela de números organizados em linhas e colunas — um
DataFrame inteiro é, matematicamente, uma matriz. Operações fundamentais: **soma** (elemento a
elemento, matrizes de mesmo tamanho), **multiplicação de matrizes** (`@` ou `np.matmul`, diferente da
multiplicação elemento a elemento — combina linhas da primeira com colunas da segunda), **transposição**
(`.T`, troca linhas por colunas), e a **matriz identidade** (`np.eye`, o "1" das matrizes: multiplicar
por ela não muda nada).

**💻 Exemplo comentado:**

````python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Soma:\n", A + B)
print("Multiplicação de matrizes (A @ B):\n", A @ B)
print("Transposta de A:\n", A.T)
print("Matriz identidade 2x2:\n", np.eye(2))
print("A @ identidade == A?", np.allclose(A @ np.eye(2), A))
````

## Capítulo 3: Determinantes

O **determinante** de uma matriz quadrada é um único número que resume propriedades
importantes dela: se o determinante é **zero**, a matriz **não tem inversa** (é "singular") — o que
tem implicações práticas em ML, por exemplo, quando features são redundantes (uma é combinação linear
de outra) e certos cálculos (como a solução exata de regressão linear via equação normal) falham.
Calculamos com `np.linalg.det()`.

**💻 Exemplo comentado:**

````python
import numpy as np

A = np.array([[1, 2], [3, 4]])
det_A = np.linalg.det(A)
print(f"Determinante de A: {det_A:.2f}")

# Matriz singular (determinante zero) - a segunda linha é múltipla da primeira
B = np.array([[1, 2], [2, 4]])
det_B = np.linalg.det(B)
print(f"Determinante de B: {det_B:.2f}")
print("B é invertível?", not np.isclose(det_B, 0))
````

## Capítulo 4: Sistemas Lineares

Um **sistema linear** é um conjunto de equações do tipo `a*x + b*y = c` que
queremos resolver simultaneamente. Representado em forma matricial como `Ax = b`, resolvemos com
`np.linalg.solve(A, b)` (muito mais eficiente e preciso do que calcular a inversa manualmente). Esse é
exatamente o tipo de cálculo que resolve a "equação normal" da regressão linear internamente.

**💻 Exemplo comentado:**

````python
import numpy as np

# Sistema: 2x + y = 5
#          x - y = 1
A = np.array([[2, 1], [1, -1]])
b = np.array([5, 1])

solucao = np.linalg.solve(A, b)
print(f"x = {solucao[0]:.2f}, y = {solucao[1]:.2f}")

# Verificando a solução
print("A @ solucao == b?", np.allclose(A @ solucao, b))
````

## Capítulo 5: Autovalores e Autovetores

Para uma matriz `A`, um **autovetor** `v` é um vetor especial que, ao ser
multiplicado por `A`, não muda de direção — apenas é escalado por um número, o **autovalor** `λ`
(lambda): `A @ v = λ * v`. Esse conceito é a base do **PCA** (Principal Component Analysis), técnica de
redução de dimensionalidade que o curso vai reencontrar mais adiante: os autovetores da matriz de
covariância indicam as direções de maior variância nos dados.

**💻 Exemplo comentado:**

````python
import numpy as np

A = np.array([[2, 0], [0, 3]])
autovalores, autovetores = np.linalg.eig(A)

print("Autovalores:", autovalores)
print("Autovetores:\n", autovetores)

# Verificando: A @ v deve ser igual a autovalor * v (para cada autovetor)
for i in range(len(autovalores)):
    v = autovetores[:, i]
    lado_esquerdo = A @ v
    lado_direito = autovalores[i] * v
    print(f"Autovetor {i}: A@v == λ*v?", np.allclose(lado_esquerdo, lado_direito))
````

## Capítulo 6: Espaços Vetoriais

Um conjunto de vetores é **linearmente independente** se nenhum deles pode ser
escrito como combinação dos outros (nenhuma informação é redundante). O **rank** (posto) de uma
matriz (`np.linalg.matrix_rank`) conta quantas linhas/colunas são realmente independentes — se o rank
é menor que o número de colunas, há redundância nas features (o que pode causar os problemas de matriz
singular vistos no Capítulo 3). A **dimensão** de um espaço vetorial é o número de vetores
independentes necessários para descrevê-lo completamente (a "base").

**💻 Exemplo comentado:**

````python
import numpy as np

# Matriz com colunas independentes
A = np.array([[1, 0], [0, 1], [1, 1]])
print("Rank de A (colunas independentes):", np.linalg.matrix_rank(A))

# Matriz com uma coluna redundante (a segunda é o dobro da primeira)
B = np.array([[1, 2], [2, 4], [3, 6]])
print("Rank de B (com redundância):", np.linalg.matrix_rank(B))
print("Número de colunas de B:", B.shape[1])
print("Há redundância?", np.linalg.matrix_rank(B) < B.shape[1])
````

````python
# --- Exercício 1 ---
import numpy as np
A = np.array([[4, 0], [0, 1]])
autovalores, autovetores = np.linalg.eig(A)
print(autovalores)

# --- Exercício 2 ---
import numpy as np
A = np.array([[2, 0], [0, 5]])
autovalores, autovetores = np.linalg.eig(A)
v = autovetores[:, 0]
print(np.allclose(A @ v, autovalores[0] * v))

# --- Exercício 3 ---
import numpy as np
A = np.array([[2, 1], [1, 2]])
autovalores, _ = np.linalg.eig(A)
print(sorted(autovalores.round(2)))

# --- Exercício 4 ---
import numpy as np
A = np.array([[3, 1], [0, 2]])
autovalores, _ = np.linalg.eig(A)
soma_autovalores = autovalores.sum()
traco = np.trace(A)
print(np.isclose(soma_autovalores, traco))

# --- Exercício 5 ---
import numpy as np
A = np.array([[2, 0], [0, 3]])
autovalores, autovetores = np.linalg.eig(A)
print(autovalores)

# --- Exercício 6 ---
import numpy as np
cov = np.array([[4, 2], [2, 3]])
autovalores, autovetores = np.linalg.eig(cov)
print("Autovalores (variância capturada):", autovalores.round(2))
print("Autovetores (direções principais):\n", autovetores.round(2))
````

2. Comparar o resultado com `sklearn.linear_model.LinearRegression`.
3. Implementar PCA simplificado: calcular a matriz de covariância e seus autovalores/autovetores.
4. Identificar a direção (autovetor) de maior variância.

### ✅ Gabarito do projeto

````python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Equação normal da regressão linear: beta = (X^T X)^-1 X^T y
X = np.array([[1, 1], [1, 2], [1, 3], [1, 4], [1, 5]])  # primeira coluna = intercepto (1s)
y = np.array([3, 5, 7, 9, 11])

Xt = X.T
beta = np.linalg.inv(Xt @ X) @ Xt @ y
print("1. Coeficientes via equação normal (intercepto, inclinação):", beta.round(3))

# 2. Comparando com scikit-learn
X_sklearn = X[:, 1:].reshape(-1, 1)  # sem a coluna de 1s, o sklearn adiciona o intercepto sozinho
modelo = LinearRegression()
modelo.fit(X_sklearn, y)
print(f"2. Scikit-learn -> intercepto: {modelo.intercept_:.3f}, coeficiente: {modelo.coef_[0]:.3f}")
print("   Resultados batem?", np.isclose(beta[0], modelo.intercept_) and np.isclose(beta[1], modelo.coef_[0]))

# 3. PCA simplificado: matriz de covariância e autovalores/autovetores
np.random.seed(0)
dados = np.random.multivariate_normal(mean=[0, 0], cov=[[3, 2], [2, 2]], size=200)
matriz_covariancia = np.cov(dados.T)
autovalores, autovetores = np.linalg.eig(matriz_covariancia)
print("\n3. Matriz de covariância:\n", matriz_covariancia.round(2))
print("   Autovalores:", autovalores.round(2))

# 4. Direção de maior variância (o primeiro componente principal)
indice_maior_autovalor = np.argmax(autovalores)
componente_principal = autovetores[:, indice_maior_autovalor]
print(f"\n4. Direção de maior variância (componente principal): {componente_principal.round(3)}")
````
