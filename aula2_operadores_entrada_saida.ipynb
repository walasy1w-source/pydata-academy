## 📘 Teoria dialogada (30 min)

**— Professor(a):** Vamos conversar sobre operadores. Você já usa eles todos os dias sem perceber:
quando divide uma pizza entre amigos, está fazendo uma divisão; quando compara preços no mercado,
está comparando valores. Python só formaliza isso com símbolos.

**— Aluno(a):** E quais símbolos são esses?

**— Professor(a):** Os **aritméticos** fazem contas: `+ - * / // % **`. Repare em dois que costumam
confundir: `//` é a **divisão inteira** (descarta a parte decimal) e `%` é o **módulo** (o resto da
divisão) — muito usado para descobrir se um número é par (`n % 2 == 0`). Já `**` é potência: `2 ** 3`
é 2 elevado a 3, ou seja, 8.

**— Aluno(a):** E os de comparação?

**— Professor(a):** `== != > < >= <=` — todos retornam `True` ou `False`. Cuidado especial com `==`
(compara) vs `=` (atribui um valor) — é o erro mais comum de quem está começando.

**— Aluno(a):** Isso me lembra o capítulo passado, sobre pegar dados do usuário...

**— Professor(a):** Exatamente, e é aí que mora uma pegadinha: `input()` **sempre** devolve texto
(`str`), mesmo que a pessoa digite `"42"`. Se você tentar somar isso com um número sem converter
primeiro (`int()` ou `float()`), o Python reclama. E para mostrar o resultado de forma bonita, usamos
**f-strings**: `f"O resultado é {resultado}"`, podendo até formatar casas decimais com
`f"{valor:.2f}"`.


**💻 Exemplo comentado:**

````python
a = 10
b = 3

print(a + b, a - b, a * b, a / b)   # aritméticos básicos
print(a // b)    # divisão inteira -> 3
print(a % b)     # resto da divisão (módulo) -> 1
print(a ** b)    # potência -> 1000

nome = input("Qual é o seu nome? ")          # sempre retorna string
idade = int(input("Qual é a sua idade? "))   # convertido para int

print(f"Olá, {nome}! Ano que vem você terá {idade + 1} anos.")
print(f"Divisão com 2 casas decimais: {a / b:.2f}")
````

## 🧑‍💻 Prática guiada (30 min) — Calculadora interativa

Vamos construir, passo a passo, uma calculadora que recebe dois números e uma operação, e devolve o
resultado formatado. Em uso interativo, os valores viriam de `input()` — aqui usamos valores fixos
para o notebook rodar sozinho, mas o comentário mostra como seria a versão interativa.

````python
# Passo 1: capturar os dois números
# Em uso interativo: num1 = float(input("Primeiro número: "))
num1 = 10.0  # valor fixo para a prática guiada
# Em uso interativo: num2 = float(input("Segundo número: "))
num2 = 3.0   # valor fixo para a prática guiada

print(f"Números capturados: {num1} e {num2}")
````

````python
# Passo 2: capturar a operação desejada
# Em uso interativo: operacao = input("Operação (+, -, *, /): ")
operacao = "+"  # valor fixo para a prática guiada

print(f"Operação escolhida: {operacao}")
````

````python
# Passo 3: aplicar o operador certo e formatar o resultado
if operacao == "+":
    resultado = num1 + num2
elif operacao == "-":
    resultado = num1 - num2
elif operacao == "*":
    resultado = num1 * num2
elif operacao == "/":
    if num2 == 0:
        resultado = None
        print("⚠️ Não é possível dividir por zero!")
    else:
        resultado = num1 / num2
else:
    resultado = None
    print("⚠️ Operação inválida!")

if resultado is not None:
    print(f"{num1} {operacao} {num2} = {resultado:.2f}")
````

## 📌 Checkpoint — Comandos da Aula 2

| Comando / Conceito | O que faz |
|---|---|
| `+ - * /` | Operadores aritméticos básicos |
| `// e %` | Divisão inteira e resto da divisão (módulo) |
| `**` | Potência |
| `== != > < >= <=` | Operadores de comparação (retornam True/False) |
| `and / or / not` | Operadores lógicos |
| `in` | Verifica se um valor está em uma coleção |
| `input()` | Captura texto digitado pelo usuário (sempre retorna string) |
| `int() / float()` | Converte texto para número inteiro/decimal |
| `f"...{var}..."` | f-string: formata texto com variáveis |
| `f"...{var:.2f}..."` | f-string com 2 casas decimais |

### ✅ Aula 2 — Soluções comentadas

````python
# --- Exercício 1 ---
print(17 % 4)  # resultado: 1

# --- Exercício 2 ---
numero = 15
print(numero % 2 == 0)  # False, 15 é ímpar

# --- Exercício 3 ---
estoque = ["arroz", "feijão", "macarrão"]
print("feijão" in estoque)

# --- Exercício 4 ---
print("O resultado é " + str(42))
# ou, de forma mais moderna, com f-string:
print(f"O resultado é {42}")

# --- Exercício 5 ---
# nome = input("Seu nome: ")
nome = "Marina"  # valor fixo para o gabarito
print(f"Olá, {nome}!")

# --- Exercício 6 ---
# a = float(input("Primeiro número: "))
# b = float(input("Segundo número: "))
a, b = 12.5, 7.5  # valores fixos para o gabarito
print(f"A soma é {a + b}")

# --- Exercício 7 ---
pacientes = [
    {"nome": "Carlos", "peso": 82, "altura": 1.75},
    {"nome": "Bia", "peso": 60, "altura": 1.65},
    {"nome": "Rafael", "peso": 95, "altura": 1.80},
]
for p in pacientes:
    imc = p["peso"] / p["altura"] ** 2
    print(f"{p['nome']}: IMC = {imc:.2f}")

# --- Exercício 8 ---
preco = 149.9
desconto = preco * 0.15
preco_final = preco - desconto
print(f"R$ {preco_final:.2f}")

# --- Desafio 1 ---
def calculadora(num1, num2, operacao):
    if operacao == "+":
        return num1 + num2
    elif operacao == "-":
        return num1 - num2
    elif operacao == "*":
        return num1 * num2
    elif operacao == "/":
        if num2 == 0:
            print("⚠️ Não é possível dividir por zero!")
            return None
        return num1 / num2
    else:
        print("⚠️ Operação inválida!")
        return None

print(calculadora(10, 2, "/"))
print(calculadora(10, 0, "/"))

# --- Desafio 2 ---
notas = [7.0, 8.5, 6.0]
pesos = [2, 3, 5]
media_ponderada = sum(n * p for n, p in zip(notas, pesos)) / sum(pesos)
print(f"Média ponderada: {media_ponderada:.2f}")
````
