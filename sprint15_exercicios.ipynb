## Capítulo 1: Fundamentos de Imagens Digitais

Uma imagem digital é uma **grade de pixels** (uma matriz numérica). Uma imagem em
**tons de cinza** tem um único canal, com valores de 0 (preto) a 255 (branco). Uma imagem **colorida**
(RGB) tem 3 canais (vermelho, verde, azul) — logo, sua forma (`shape`) é `(altura, largura, 3)`. Em
Python, representamos imagens como arrays NumPy: `imagem.shape` dá as dimensões, `imagem[y, x]` acessa
um pixel específico.

**💻 Exemplo comentado:**

````python
import numpy as np

# Imagem em tons de cinza 5x5 (valores de 0 a 255)
imagem_cinza = np.array([
    [0, 50, 100, 150, 200],
    [50, 100, 150, 200, 250],
    [100, 150, 200, 250, 255],
    [150, 200, 250, 255, 255],
    [200, 250, 255, 255, 255],
], dtype=np.uint8)

print("Shape (altura, largura):", imagem_cinza.shape)
print("Valor do pixel (2,2):", imagem_cinza[2, 2])
print("Valor médio de brilho:", imagem_cinza.mean())

# Imagem colorida (RGB) 3x3 - shape (altura, largura, canais)
imagem_rgb = np.zeros((3, 3, 3), dtype=np.uint8)
imagem_rgb[0, 0] = [255, 0, 0]   # pixel vermelho puro
imagem_rgb[1, 1] = [0, 255, 0]   # pixel verde puro
print("\nShape da imagem RGB:", imagem_rgb.shape)
print("Pixel (0,0) - vermelho puro:", imagem_rgb[0, 0])
````

## Capítulo 2: Processamento de Imagens com OpenCV

O **OpenCV** (`cv2`) é a biblioteca padrão para processamento de imagens em Python.
Operações comuns: `cv2.cvtColor` (converter entre espaços de cor, ex.: RGB para tons de cinza),
`cv2.resize` (redimensionar), `cv2.GaussianBlur` (desfoque, reduz ruído), `cv2.threshold`
(binarização: transforma em preto/branco puro com base em um limiar), e funções de desenho como
`cv2.rectangle`/`cv2.circle`. O OpenCV opera diretamente sobre arrays NumPy.

**💻 Exemplo comentado:**

````python
import cv2
import numpy as np

# Imagem colorida sintética (100x100, fundo cinza)
imagem = np.full((100, 100, 3), 150, dtype=np.uint8)
cv2.rectangle(imagem, (20, 20), (80, 80), (0, 0, 255), -1)  # retângulo vermelho preenchido

# Convertendo para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
print("Shape original:", imagem.shape, "-> Shape em cinza:", imagem_cinza.shape)

# Redimensionando
imagem_redimensionada = cv2.resize(imagem, (50, 50))
print("Shape redimensionado:", imagem_redimensionada.shape)

# Binarização (threshold)
_, imagem_binaria = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)
print("Valores únicos após threshold:", np.unique(imagem_binaria))
````

## Capítulo 3: Redes Neurais Convolucionais (CNNs)

CNNs são a arquitetura por trás da maioria dos avanços em visão computacional. A
operação central é a **convolução**: um pequeno **filtro/kernel** (ex.: 3x3) "desliza" sobre a imagem,
multiplicando e somando valores em cada posição — diferentes kernels detectam diferentes padrões
(bordas, texturas). O **pooling** (ex.: max pooling) reduz o tamanho da imagem resultante, mantendo
as informações mais importantes. Vamos implementar essas operações manualmente com NumPy para entender
a intuição, sem precisar de um framework de deep learning completo.

**💻 Exemplo comentado:**

````python
import numpy as np

def convolucao_2d(imagem, kernel):
    altura_k, largura_k = kernel.shape
    altura_img, largura_img = imagem.shape
    saida = np.zeros((altura_img - altura_k + 1, largura_img - largura_k + 1))
    for i in range(saida.shape[0]):
        for j in range(saida.shape[1]):
            regiao = imagem[i:i+altura_k, j:j+largura_k]
            saida[i, j] = np.sum(regiao * kernel)
    return saida

# Kernel de detecção de borda vertical
kernel_borda = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

imagem = np.array([
    [0, 0, 100, 100, 100],
    [0, 0, 100, 100, 100],
    [0, 0, 100, 100, 100],
    [0, 0, 100, 100, 100],
], dtype=float)

resultado = convolucao_2d(imagem, kernel_borda)
print("Resultado da convolução (detecção de borda):")
print(resultado)

def max_pooling(imagem, tamanho=2):
    altura, largura = imagem.shape
    saida = np.zeros((altura // tamanho, largura // tamanho))
    for i in range(0, altura - tamanho + 1, tamanho):
        for j in range(0, largura - tamanho + 1, tamanho):
            saida[i // tamanho, j // tamanho] = np.max(imagem[i:i+tamanho, j:j+tamanho])
    return saida

print("\nApós max pooling 2x2:")
print(max_pooling(resultado))
````

## Capítulo 4: Classificação de Imagens

Para classificar imagens com técnicas de ML "clássicas" (sem treinar uma CNN
completa, que exige mais dados e tempo), uma abordagem simples e didática é **achatar (flatten)** cada
imagem em um vetor de pixels e usar esse vetor como features de um classificador comum (Sprint 8/9),
como `LogisticRegression` ou `SVC`. Não captura padrões espaciais tão bem quanto uma CNN, mas ilustra
o fluxo geral: imagem → vetor de features → classificador.

**💻 Exemplo comentado:**

````python
import numpy as np
from sklearn.svm import SVC

def gerar_imagem_quadrado(tamanho=8, brilho_fundo=20, brilho_quadrado=200):
    imagem = np.full((tamanho, tamanho), brilho_fundo, dtype=float)
    meio = tamanho // 2
    imagem[meio-1:meio+2, meio-1:meio+2] = brilho_quadrado
    return imagem

def gerar_imagem_uniforme(tamanho=8, brilho=100):
    return np.full((tamanho, tamanho), brilho, dtype=float)

# Gerando um pequeno dataset: "tem quadrado brilhante" (1) vs "uniforme" (0)
imagens = [gerar_imagem_quadrado(), gerar_imagem_quadrado(brilho_quadrado=180),
           gerar_imagem_uniforme(), gerar_imagem_uniforme(brilho=90)]
rotulos = [1, 1, 0, 0]

X = np.array([img.flatten() for img in imagens])   # achatando cada imagem em um vetor
print("Shape de X (imagens achatadas):", X.shape)

modelo = SVC(kernel="linear")
modelo.fit(X, rotulos)

nova_imagem = gerar_imagem_quadrado(brilho_quadrado=220).flatten()
print("Previsão para nova imagem com quadrado:", modelo.predict([nova_imagem]))
````

## Capítulo 5: Detecção de Objetos

Diferente da classificação (que diz "o que" está na imagem inteira), a **detecção de
objetos** localiza **onde** um objeto está, geralmente com uma **bounding box** (caixa delimitadora:
coordenadas do canto superior esquerdo e do canto inferior direito). Modelos de produção usam
arquiteturas como YOLO ou Faster R-CNN — aqui, ilustramos a intuição com uma versão simplificada:
localizar a região de maior intensidade (ou acima de um limiar) em uma imagem sintética usando
`np.where`.

**💻 Exemplo comentado:**

````python
import numpy as np

def encontrar_bounding_box(imagem, limiar=150):
    linhas, colunas = np.where(imagem > limiar)
    if len(linhas) == 0:
        return None
    return {
        "topo": int(linhas.min()),
        "base": int(linhas.max()),
        "esquerda": int(colunas.min()),
        "direita": int(colunas.max()),
    }

imagem = np.array([
    [10, 10, 10, 10, 10],
    [10, 200, 200, 200, 10],
    [10, 200, 200, 200, 10],
    [10, 10, 10, 10, 10],
])

bbox = encontrar_bounding_box(imagem)
print("Bounding box do objeto detectado:", bbox)
````

````python
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def gerar_imagem_bem_enquadrada(tamanho=8, semente=0):
    rng = np.random.RandomState(semente)
    imagem = rng.uniform(0, 40, (tamanho, tamanho))   # fundo escuro
    meio = tamanho // 2
    imagem[meio-2:meio+2, meio-2:meio+2] = rng.uniform(200, 255, (4, 4))  # objeto claro centralizado
    return imagem

def gerar_imagem_mal_enquadrada(tamanho=8, semente=0):
    rng = np.random.RandomState(semente)
    return rng.uniform(80, 160, (tamanho, tamanho))   # sem contraste claro, tudo "médio"

# 1. Dataset sintético
imagens_boas = [gerar_imagem_bem_enquadrada(semente=i) for i in range(15)]
imagens_ruins = [gerar_imagem_mal_enquadrada(semente=i) for i in range(15)]
imagens = imagens_boas + imagens_ruins
rotulos = [1] * 15 + [0] * 15
print(f"1. Dataset gerado com {len(imagens)} imagens.")

# 2. Achatando e treinando
X = np.array([img.flatten() for img in imagens])
y = np.array(rotulos)
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

modelo = SVC(kernel="linear")
modelo.fit(X_treino, y_treino)
print("2. Modelo treinado.")

# 3. Avaliação
acuracia = accuracy_score(y_teste, modelo.predict(X_teste))
print(f"3. Acurácia no teste: {acuracia:.2f}")

# 4. Localizando o objeto (bounding box) em uma imagem bem enquadrada
def encontrar_bounding_box(imagem, limiar=150):
    linhas, colunas = np.where(imagem > limiar)
    if len(linhas) == 0:
        return None
    return {"topo": int(linhas.min()), "base": int(linhas.max()), "esquerda": int(colunas.min()), "direita": int(colunas.max())}

imagem_teste = gerar_imagem_bem_enquadrada(semente=99)
bbox = encontrar_bounding_box(imagem_teste)
print(f"4. Bounding box do objeto detectado: {bbox}")
````
