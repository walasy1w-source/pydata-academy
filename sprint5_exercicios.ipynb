## Capítulo 1: Controle de Versão com Git e GitHub

**Git** é um sistema de controle de versão: ele grava um "histórico de fotos" do seu
código ao longo do tempo, permitindo voltar, comparar e colaborar sem medo de perder nada. Os comandos
essenciais: `git init` (cria um repositório), `git add` (marca arquivos para o próximo commit), `git
commit -m "mensagem"` (salva uma versão), `git log` (mostra o histórico), `git branch`/`git checkout
-b` (cria e alterna linhas paralelas de desenvolvimento), `git merge` (junta branches). O **GitHub** é
um serviço online que hospeda repositórios Git e permite colaboração (`git push`, `git pull`, `git
clone`, Pull Requests).

📌 Neste notebook, os exemplos criam repositórios Git **locais e temporários** via `subprocess`, para
que você veja o comportamento real dos comandos sem precisar de internet ou de uma conta no GitHub.

**💻 Exemplo comentado:**

````python
import subprocess
import tempfile
import os

def rodar_git(args, cwd):
    resultado = subprocess.run(["git"] + args, cwd=cwd, capture_output=True, text=True)
    return resultado.stdout.strip()

pasta = tempfile.mkdtemp()
rodar_git(["init"], pasta)
rodar_git(["config", "user.email", "aluno@pydata.academy"], pasta)
rodar_git(["config", "user.name", "Aluno PyData"], pasta)

with open(os.path.join(pasta, "ola.py"), "w") as f:
    f.write("print('Olá, Git!')\n")

rodar_git(["add", "."], pasta)
rodar_git(["commit", "-m", "Primeiro commit"], pasta)

log = rodar_git(["log", "--oneline"], pasta)
print(log)
````

## Capítulo 3: Boas Práticas e PEP 8

A **PEP 8** é o guia de estilo oficial do Python. Principais regras: nomes de
variáveis e funções em `snake_case` (não `camelCase` ou `PascalCase`, reservados para classes), linhas
com no máximo 79-99 caracteres, espaço após vírgulas, dois espaços em branco antes de comentários
inline, e nomes descritivos (`total_vendas` em vez de `tv`). Código que segue PEP 8 é mais fácil de
ler, revisar e manter em equipe.

**💻 Exemplo comentado:**

````python
# ❌ Fora do padrão PEP 8
def CalculaMedia(Lista):
    Soma=0
    for x in Lista:Soma+=x
    return Soma/len(Lista)

# ✅ Seguindo a PEP 8
def calcula_media(lista):
    soma = 0
    for numero in lista:
        soma += numero
    return soma / len(lista)

print(calcula_media([10, 20, 30]))
````

## Capítulo 4: Testes Automatizados com Pytest

Testes automatizados são funções que verificam se seu código se comporta como
esperado, usando `assert`. O **Pytest** é o framework de testes mais popular em Python: qualquer
função cujo nome comece com `test_` em um arquivo `test_*.py` é executada automaticamente pelo comando
`pytest`. Dentro dela, usamos `assert condicao` — se a condição for falsa, o teste falha e o Pytest
mostra exatamente o que esperava e o que recebeu.

📌 Neste notebook, escrevemos funções `test_*` no estilo Pytest e as executamos diretamente com uma
chamada de função — o comportamento (assert falha = teste falha) é idêntico ao que o Pytest faria.

**💻 Exemplo comentado:**

````python
def soma(a, b):
    return a + b

def test_soma_numeros_positivos():
    assert soma(2, 3) == 5

def test_soma_numeros_negativos():
    assert soma(-1, -1) == -2

# "Rodando" os testes manualmente (o Pytest faria isso automaticamente)
test_soma_numeros_positivos()
test_soma_numeros_negativos()
print("Todos os testes passaram!")
````

## Capítulo 5: Depuração (Debugging)

Depurar é o processo de encontrar e corrigir erros. As técnicas mais usadas: **ler o
traceback com calma** (a última linha diz o tipo de erro, as anteriores mostram o caminho até ele),
**print debugging** (inserir `print()` para ver valores intermediários), e o depurador `pdb`
(`import pdb; pdb.set_trace()`, que pausa a execução e permite inspecionar variáveis). A regra de ouro:
isole o problema reduzindo o código até o menor trecho que ainda reproduz o erro.

**💻 Exemplo comentado:**

````python
def calcular_media(notas):
    return sum(notas) / len(notas)

notas_turma = [7, 8, 6, 9]
print(f"Média: {calcular_media(notas_turma)}")

# Debugging com print: inspecionando um valor suspeito antes de usá-lo
notas_vazias = []
if len(notas_vazias) == 0:
    print("Lista vazia detectada — evitando ZeroDivisionError")
else:
    print(calcular_media(notas_vazias))
````

## Capítulo 6: Documentação de Código

Código bem documentado é mais fácil de manter e de usar por outras pessoas (inclusive
você mesmo, no futuro). As **docstrings** (`"""texto"""` logo após a definição de uma função)
descrevem o que ela faz, seus parâmetros e o retorno, e ficam acessíveis via `função.__doc__` ou
`help(função)`. **Type hints** (`def soma(a: int, b: int) -> int:`) documentam os tipos esperados sem
forçá-los em tempo de execução. Um bom `README.md` documenta o projeto inteiro: o que é, como instalar,
como usar.

**💻 Exemplo comentado:**

````python
def calcular_imc(peso: float, altura: float) -> float:
    """
    Calcula o Índice de Massa Corporal (IMC).

    Args:
        peso: peso em quilogramas.
        altura: altura em metros.

    Returns:
        O IMC calculado (peso / altura ao quadrado).
    """
    return peso / (altura ** 2)

print(calcular_imc(70, 1.75))
print(calcular_imc.__doc__)
print(calcular_imc.__annotations__)
````

## Capítulo 7: Linha de Comando (Terminal)

O terminal é a forma mais direta de interagir com o computador: `pwd` (mostra a pasta
atual), `ls`/`dir` (lista arquivos), `cd` (muda de pasta), `mkdir` (cria pasta), `cp`/`mv`/`rm` (copia,
move, remove), `cat` (mostra conteúdo de um arquivo), `grep` (busca texto). Em Python, podemos executar
esses comandos programaticamente com o módulo `subprocess` — útil para automatizar tarefas.

**💻 Exemplo comentado:**

````python
import subprocess
import tempfile
import os

pasta = tempfile.mkdtemp()

# Criando arquivos com Python (equivalente a "touch")
for nome in ["a.txt", "b.txt", "c.py"]:
    with open(os.path.join(pasta, nome), "w") as f:
        f.write("conteúdo de teste")

# "ls" via subprocess
resultado = subprocess.run(["ls", pasta], capture_output=True, text=True)
print(resultado.stdout)

# "grep" buscando arquivos .py
arquivos_py = [linha for linha in resultado.stdout.splitlines() if linha.endswith(".py")]
print(arquivos_py)
````

## Capítulo 8: Trabalhando com APIs

um objeto `response`, e `response.json()` converte o corpo da resposta em um dicionário/lista Python.
O `response.status_code` informa se deu certo (`200` = sucesso, `404` = não encontrado, `500` = erro no
servidor).
