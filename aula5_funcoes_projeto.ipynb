## 📘 Teoria dialogada (30 min)

**— Aluno(a):** Depois de aprender tanta coisa solta, como eu organizo meu código?

**— Professor(a):** Com **funções**! Pense em uma função como uma máquina de suco: você coloca frutas
(os **parâmetros**) e recebe um suco pronto (o **retorno**, com `return`). Isso evita repetir o mesmo
código várias vezes.

**— Aluno(a):** E os parâmetros sempre são obrigatórios?

**— Professor(a):** Não! Você pode dar um **valor padrão**: `def saudacao(nome, cumprimento="Olá"):` —
se ninguém informar o cumprimento, ele usa "Olá". E se você não souber quantos argumentos vão chegar,
usa `*args`: `def somar_tudo(*numeros): return sum(numeros)` aceita qualquer quantidade de números.

**— Aluno(a):** E o que é uma função lambda?

**— Professor(a):** É uma função "descartável", de uma linha só, sem nome: `lambda n: n * 2`. Ela
brilha quando usada com `map()` (aplica uma função a cada item de uma lista) e `filter()` (mantém só
os itens que passam em um teste) — uma primeira porta de entrada para a **programação funcional**,
um estilo muito comum em Ciência de Dados.

**— Aluno(a):** Isso tudo vai servir para o quê hoje?

**— Professor(a):** Para fechar o Sprint 1 com um projeto de verdade: analisar as vendas de uma loja
fictícia, a **PyData Store**, usando exatamente essas ferramentas.


**💻 Exemplo comentado:**

````python
def saudacao(nome, cumprimento="Olá"):
    return f"{cumprimento}, {nome}!"

print(saudacao("Pedro"))
print(saudacao("Ana", "Bom dia"))

def somar_tudo(*numeros):
    return sum(numeros)

print(somar_tudo(1, 2, 3, 4))  # *args aceita qualquer quantidade de argumentos

numeros = [1, 2, 3, 4, 5]
quadrados = list(map(lambda n: n ** 2, numeros))     # aplica a função a cada item
pares = list(filter(lambda n: n % 2 == 0, numeros))  # mantém só os que passam no teste
print(quadrados)
print(pares)
````

## 🧑‍💻 Prática guiada (30 min) — Funções para análise de vendas

Vamos construir, peça por peça, as funções que usaremos no projeto integrador logo a seguir.

````python
vendas_exemplo = [120, 340, 95, 210, 180]

def total_vendas(vendas):
    return sum(vendas)

def media_vendas(vendas):
    return sum(vendas) / len(vendas)

def maior_venda(vendas):
    return max(vendas)

def menor_venda(vendas):
    return min(vendas)

print(f"Total: {total_vendas(vendas_exemplo)}")
print(f"Média: {media_vendas(vendas_exemplo):.2f}")
print(f"Maior: {maior_venda(vendas_exemplo)}")
print(f"Menor: {menor_venda(vendas_exemplo)}")
````

````python
# Agora combinamos tudo em uma única função que retorna um dicionário:
def resumo_vendas(vendas):
    return {
        "total": total_vendas(vendas),
        "media": media_vendas(vendas),
        "maximo": maior_venda(vendas),
        "minimo": menor_venda(vendas),
    }

print(resumo_vendas(vendas_exemplo))
````

### ✅ Gabarito do projeto integrador

````python
vendas = [
    {"produto": "Caneta",    "quantidade": 50, "preco_unitario": 2.5},
    {"produto": "Caderno",   "quantidade": 20, "preco_unitario": 15.0},
    {"produto": "Caneta",    "quantidade": 30, "preco_unitario": 2.5},
    {"produto": "Mochila",   "quantidade": 5,  "preco_unitario": 120.0},
    {"produto": "Caderno",   "quantidade": 12, "preco_unitario": 15.0},
    {"produto": "Lápis",     "quantidade": 80, "preco_unitario": 1.2},
    {"produto": "Mochila",   "quantidade": 3,  "preco_unitario": 120.0},
    {"produto": "Estojo",    "quantidade": 18, "preco_unitario": 22.0},
]

def resumo_vendas(vendas):
    """Retorna um dicionário com total, media, maximo e minimo do valor de cada venda."""
    valores = [v["quantidade"] * v["preco_unitario"] for v in vendas]
    return {
        "total": sum(valores),
        "media": sum(valores) / len(valores),
        "maximo": max(valores),
        "minimo": min(valores),
    }

def produto_mais_vendido(vendas):
    """Retorna o nome do produto com a maior quantidade total vendida."""
    quantidade_por_produto = {}
    for v in vendas:
        produto = v["produto"]
        quantidade_por_produto[produto] = quantidade_por_produto.get(produto, 0) + v["quantidade"]
    return max(quantidade_por_produto, key=quantidade_por_produto.get)

resumo = resumo_vendas(vendas)
mais_vendido = produto_mais_vendido(vendas)

print("📊 Relatório de Vendas — PyData Store")
print("-" * 40)
print(f"Faturamento total: R$ {resumo['total']:.2f}")
print(f"Ticket médio por venda: R$ {resumo['media']:.2f}")
print(f"Maior venda individual: R$ {resumo['maximo']:.2f}")
print(f"Menor venda individual: R$ {resumo['minimo']:.2f}")
print(f"Produto mais vendido (por quantidade): {mais_vendido}")
````

### 💪 Desafio bônus — gráfico de barras com matplotlib

Requer a biblioteca `matplotlib` instalada (`pip install matplotlib`). Se ela não estiver disponível
no seu ambiente, este é o único bloco do notebook que pode falhar — os demais não dependem dela.

````python
import matplotlib.pyplot as plt

quantidade_por_produto = {}
for v in vendas:
    produto = v["produto"]
    quantidade_por_produto[produto] = quantidade_por_produto.get(produto, 0) + v["quantidade"]

produtos = list(quantidade_por_produto.keys())
quantidades = list(quantidade_por_produto.values())

plt.figure(figsize=(8, 4))
plt.bar(produtos, quantidades, color="#d4af37")
plt.title("Quantidade vendida por produto — PyData Store")
plt.xlabel("Produto")
plt.ylabel("Quantidade vendida")
plt.tight_layout()
plt.show()
print("✅ Gráfico gerado com sucesso!")
````
