## Capítulo 1: Definição do Problema

Todo projeto de dados começa com uma pergunta — e a qualidade da pergunta define a
qualidade do projeto inteiro. Uma pergunta vaga ("por que perdemos clientes?") precisa virar uma
**pergunta de negócio bem definida**: quem é o público, qual período, qual métrica de sucesso.
Definimos também **KPIs** (indicadores-chave) que vamos medir, e o **escopo** (o que está dentro e
fora do projeto).

📌 Regra prática: uma boa pergunta de negócio tem um verbo de ação, um público-alvo claro e um período
de tempo. Exemplo ruim: "Melhorar o produto". Exemplo bom: "Reduzir o churn de assinantes do plano
mensal em 10% nos próximos 3 meses".

**💻 Exemplo comentado:**

````python
# Transformando uma pergunta vaga em uma pergunta de negócio bem definida

pergunta_vaga = "Por que estamos perdendo clientes?"

pergunta_bem_definida = {
    "pergunta": "Quais fatores mais influenciam o cancelamento (churn) de assinantes do plano mensal?",
    "publico_alvo": "Assinantes do plano mensal nos últimos 6 meses",
    "metrica_sucesso": "Taxa de churn mensal (%)",
    "escopo": "Não inclui assinantes de planos anuais nem clientes corporativos",
}

for chave, valor in pergunta_bem_definida.items():
    print(f"{chave}: {valor}")
````

## Capítulo 3: Análise Exploratória

Com dados limpos em mãos, a **análise exploratória de dados (EDA)** busca padrões:
estatísticas descritivas por grupo, correlações entre variáveis e visualizações que revelem tendências.
O objetivo aqui não é confirmar uma teoria, mas **deixar os dados "falarem"** antes de tirar conclusões
precipitadas.

**💻 Exemplo comentado:**

````python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 200
dados = pd.DataFrame({
    "meses_de_assinatura": np.random.randint(1, 36, n),
    "horas_assistidas_mes": np.random.normal(20, 8, n).round(1),
    "reclamacoes_suporte": np.random.poisson(0.5, n),
    "cancelou": np.random.choice([0, 1], n, p=[0.75, 0.25]),
})

# Estatística descritiva por grupo (cancelou vs não cancelou)
resumo = dados.groupby("cancelou")[["horas_assistidas_mes", "reclamacoes_suporte"]].mean()
print(resumo)

# Correlação entre reclamações e cancelamento
correlacao = dados["reclamacoes_suporte"].corr(dados["cancelou"])
print(f"\nCorrelação reclamações x cancelamento: {correlacao:.3f}")
````

## Capítulo 4: Construção de Relatório

Um bom relatório de dados traduz números em **decisões**. A estrutura clássica:
contexto/problema, principais achados (com números concretos), e recomendações acionáveis. Evite
jargão técnico desnecessário — quem lê o relatório muitas vezes não sabe o que é um "p-value" ou um
"DataFrame", mas entende perfeitamente "clientes com mais de 2 reclamações têm 3x mais chance de
cancelar".

**💻 Exemplo comentado:**

````python
achados = {
    "taxa_churn_geral": 25.0,
    "reclamacoes_media_cancelaram": 1.8,
    "reclamacoes_media_ficaram": 0.4,
    "horas_media_cancelaram": 9.2,
    "horas_media_ficaram": 22.5,
}

relatorio = f"""
RELATÓRIO — Análise de Churn de Assinantes (PyData Streaming)

Contexto: {achados['taxa_churn_geral']}% dos assinantes do plano mensal cancelaram no período analisado.

Principais achados:
- Assinantes que cancelaram tinham em média {achados['reclamacoes_media_cancelaram']} reclamações de
  suporte, contra {achados['reclamacoes_media_ficaram']} de quem permaneceu.
- Assinantes que cancelaram assistiam em média {achados['horas_media_cancelaram']}h/mês,
  contra {achados['horas_media_ficaram']}h/mês de quem permaneceu.

Recomendação: priorizar o atendimento de assinantes com 2+ reclamações e criar alertas para quedas
bruscas no consumo mensal — ambos são sinais precoces de risco de cancelamento.
"""
print(relatorio)
````

## Capítulo 5: Apresentação do Projeto

Apresentar um projeto de dados — seja para o time, para a liderança, ou em um
portfólio — exige uma estrutura narrativa: **problema → jornada → resultado**. Comece pelo "porquê"
(qual pergunta o projeto respondeu), mostre o caminho percorrido (dados, análise) de forma resumida, e
termine com o impacto (o que muda com esse projeto). Um portfólio de dados forte tem 3-5 projetos bem
documentados, cada um com um README explicando contexto, processo e resultado — mais valioso do que
20 projetos incompletos.

**💻 Exemplo comentado:**

````python
slides = [
    {"titulo": "O Problema", "conteudo": "25% dos assinantes do plano mensal cancelam. Por quê?"},
    {"titulo": "A Jornada", "conteudo": "Analisamos 200 assinantes: horas assistidas, reclamações, tempo de assinatura."},
    {"titulo": "O Achado Principal", "conteudo": "Reclamações de suporte e baixo consumo são os maiores sinais de risco."},
    {"titulo": "A Recomendação", "conteudo": "Priorizar suporte para quem reclama 2+ vezes e criar alertas de baixo consumo."},
]

for i, slide in enumerate(slides, start=1):
    print(f"Slide {i}: {slide['titulo']}")
    print(f"  {slide['conteudo']}\n")
````

````python
import numpy as np
import pandas as pd

# ETAPA 1: Definição do problema
projeto_churn = {
    "pergunta": "Quais fatores mais influenciam o cancelamento de assinantes do plano mensal da PyData Streaming?",
    "publico_alvo": "Assinantes do plano mensal nos últimos 6 meses",
    "metrica_sucesso": "Taxa de churn mensal (%)",
    "escopo": "Não inclui assinantes de planos anuais nem contas corporativas",
}
pergunta_projeto = projeto_churn["pergunta"]
print("ETAPA 1 — Problema definido:")
print(f"  {pergunta_projeto}\n")

# ETAPA 2: Coleta e preparação dos dados
np.random.seed(42)
n = 200
dados = pd.DataFrame({
    "assinante_id": range(1, n + 1),
    "meses_de_assinatura": np.random.randint(1, 36, n),
    "horas_assistidas_mes": np.random.normal(20, 8, n).round(1),
    "reclamacoes_suporte": np.random.poisson(0.5, n),
    "cancelou": np.random.choice([0, 1], n, p=[0.75, 0.25]),
})
dados.loc[5, "horas_assistidas_mes"] = np.nan
dados = pd.concat([dados, dados.iloc[[10]]], ignore_index=True)

dados_limpos = dados.copy()
dados_limpos["horas_assistidas_mes"] = dados_limpos["horas_assistidas_mes"].fillna(dados_limpos["horas_assistidas_mes"].mean())
dados_limpos = dados_limpos.drop_duplicates()
print(f"ETAPA 2 — Dados coletados e limpos: {dados_limpos.shape[0]} assinantes, 0 nulos: {dados_limpos.isnull().sum().sum() == 0}\n")

# ETAPA 3: Análise exploratória
resumo = dados_limpos.groupby("cancelou")[["horas_assistidas_mes", "reclamacoes_suporte"]].mean()
correlacao = dados_limpos["reclamacoes_suporte"].corr(dados_limpos["cancelou"])
print("ETAPA 3 — Análise exploratória:")
print(resumo)
print(f"  Correlação reclamações x cancelamento: {correlacao:.3f}\n")

# ETAPA 4: Construção do relatório
taxa_churn = dados_limpos["cancelou"].mean() * 100
achados = {
    "taxa_churn_geral": round(taxa_churn, 1),
    "reclamacoes_media_cancelaram": round(resumo.loc[1, "reclamacoes_suporte"], 2),
    "reclamacoes_media_ficaram": round(resumo.loc[0, "reclamacoes_suporte"], 2),
    "horas_media_cancelaram": round(resumo.loc[1, "horas_assistidas_mes"], 1),
    "horas_media_ficaram": round(resumo.loc[0, "horas_assistidas_mes"], 1),
}
taxa_churn_geral = achados["taxa_churn_geral"]
reclamacoes_media_cancelaram = achados["reclamacoes_media_cancelaram"]
reclamacoes_media_ficaram = achados["reclamacoes_media_ficaram"]
horas_media_cancelaram = achados["horas_media_cancelaram"]
horas_media_ficaram = achados["horas_media_ficaram"]

relatorio_final = f"""
RELATÓRIO — Análise de Churn de Assinantes (PyData Streaming)

Contexto: {taxa_churn_geral}% dos assinantes do plano mensal cancelaram no período analisado.

Principais achados:
- Assinantes que cancelaram tinham em média {reclamacoes_media_cancelaram} reclamações,
  contra {reclamacoes_media_ficaram} de quem permaneceu.
- Assinantes que cancelaram assistiam em média {horas_media_cancelaram}h/mês,
  contra {horas_media_ficaram}h/mês de quem permaneceu.

Recomendação: priorizar o atendimento de assinantes com 2+ reclamações e criar alertas para quedas
bruscas no consumo mensal.
"""
print("ETAPA 4 — Relatório:")
print(relatorio_final)

# ETAPA 5: Apresentação do projeto
slides_finais = [
    {"titulo": "O Problema", "conteudo": f"{taxa_churn_geral}% dos assinantes do plano mensal cancelam."},
    {"titulo": "A Jornada", "conteudo": "Analisamos 200 assinantes: horas assistidas, reclamações e tempo de assinatura."},
    {"titulo": "O Achado Principal", "conteudo": "Quem cancela reclama mais e assiste menos horas por mês."},
    {"titulo": "A Recomendação", "conteudo": "Alertas automáticos para 2+ reclamações ou queda brusca no consumo."},
]
print("ETAPA 5 — Apresentação:")
for i, slide in enumerate(slides_finais, start=1):
    titulo_slide = slide["titulo"]
    conteudo_slide = slide["conteudo"]
    print(f"  Slide {i}: {titulo_slide} — {conteudo_slide}")

print("\n✅ Projeto Integrado 1 concluído de ponta a ponta!")
````
