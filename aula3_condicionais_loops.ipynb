## 📘 Teoria dialogada (30 min)

**— Aluno(a):** Depois de aprender a fazer contas, o que vem agora?

**— Professor(a):** Agora seu programa aprende a **decidir**. O `if` testa uma condição; se for
verdadeira, executa um bloco de código. Se for falsa, o Python pode cair num `elif` (outra condição)
ou num `else` (o que sobrou). É como um fluxograma: "se está chovendo, leve guarda-chuva; senão, leve
óculos de sol".

**— Aluno(a):** E como eu repito uma ação várias vezes?

**— Professor(a):** Com **loops**. O `for` é perfeito quando você sabe quantas vezes quer repetir (ou
está percorrendo uma lista): `for i in range(5):` roda 5 vezes. Já o `while` repete **enquanto uma
condição for verdadeira** — ótimo para "repita até o usuário acertar a senha", por exemplo.

**— Aluno(a):** E se eu quiser parar um loop no meio, ou pular uma repetição?

**— Professor(a):** `break` interrompe o loop imediatamente. `continue` pula para a próxima repetição,
sem executar o resto do bloco daquela vez. E cuidado: se a condição do `while` nunca ficar falsa
(porque você esqueceu de atualizar uma variável, por exemplo), seu programa entra em **loop
infinito** — um dos bugs mais comuns (e frustrantes) para quem está aprendendo!


**💻 Exemplo comentado:**

````python
nota = 7.5

if nota >= 9:
    conceito = "Excelente"
elif nota >= 7:
    conceito = "Bom"
else:
    conceito = "Precisa melhorar"
print(f"Conceito: {conceito}")

# for com range
for i in range(1, 4):
    print(f"Repetição {i}")

# while com break/continue
contador = 0
while contador < 10:
    contador += 1
    if contador % 2 == 0:
        continue      # pula os números pares
    if contador > 7:
        break          # para de vez quando passar de 7
    print(contador)
````

### 🎯 Jogo de adivinhação (while)
Em uso interativo, o jogador digitaria palpites com `input()`. Aqui simulamos os palpites em uma
lista, para o notebook rodar sozinho.

````python
numero_secreto = 42
palpites_simulados = [10, 50, 40, 42]  # simula o que o jogador digitaria em input()
tentativa = 0
acertou = False

for palpite in palpites_simulados:
    tentativa += 1
    if palpite == numero_secreto:
        acertou = True
        break
    elif palpite < numero_secreto:
        print(f"Tentativa {tentativa}: {palpite} — é MAIOR que isso!")
    else:
        print(f"Tentativa {tentativa}: {palpite} — é MENOR que isso!")

if acertou:
    print(f"🎉 Acertou em {tentativa} tentativa(s)! O número era {numero_secreto}.")
````

### 🎯 Classificador de idade (if/elif)

````python
def classificar_idade(idade):
    if idade <= 12:
        return "criança"
    elif idade <= 17:
        return "adolescente"
    elif idade <= 59:
        return "adulto"
    else:
        return "idoso"

for idade_teste in [8, 15, 34, 70]:
    print(f"{idade_teste} anos -> {classificar_idade(idade_teste)}")
````

## 📌 Checkpoint — Comandos da Aula 3

| Comando / Conceito | O que faz |
|---|---|
| `if / elif / else` | Estruturas condicionais |

### ✅ Aula 3 — Soluções comentadas

````python
# --- Exercício 1 ---
numero = -4
if numero > 0:
    print("Positivo")
elif numero < 0:
    print("Negativo")
else:
    print("Zero")

# --- Exercício 2 ---
numero = 9
if numero % 2 == 0:
    print("Par")
else:
    print("Ímpar")

# --- Exercício 3 ---
idade = 45
if idade <= 12:
    faixa = "criança"
elif idade <= 17:
    faixa = "adolescente"
elif idade <= 59:
    faixa = "adulto"
else:
    faixa = "idoso"
print(faixa)

# --- Exercício 4 ---
x = 5
while x > 0:
    print(x)
    x -= 1  # sem esta linha, x nunca muda e o loop nunca termina

# --- Exercício 5 ---
for i in range(1, 11):
    print(i)

# --- Exercício 6 ---
soma = sum(range(1, 101))
print(soma)

# --- Exercício 7 ---
produtos = [
    {"nome": "Notebook", "preco": 3200},
    {"nome": "Mouse", "preco": 80},
    {"nome": "Monitor", "preco": 1200},
]
for p in produtos:
    desconto = 0.10 if p["preco"] > 1000 else 0.05
    preco_final = p["preco"] * (1 - desconto)
    print(f"{p['nome']}: R$ {preco_final:.2f}")

# --- Exercício 8 ---
numero = 7
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")

# --- Desafio 1 ---
for n in range(1, 31):
    if n % 15 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)

# --- Desafio 2 ---
numero_secreto = 42
palpites_simulados = [10, 50, 40, 42]
maiores, menores = 0, 0
acertou = False

for palpite in palpites_simulados:
    if palpite == numero_secreto:
        acertou = True
        break
    elif palpite < numero_secreto:
        menores += 1
    else:
        maiores += 1

print(f"Palpites menores que o número: {menores}")
print(f"Palpites maiores que o número: {maiores}")
print("Acertou!" if acertou else "Não acertou.")
````
