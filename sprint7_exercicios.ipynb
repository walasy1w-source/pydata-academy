## Capítulo 1: Introdução a Bancos de Dados Relacionais

Um **banco de dados relacional** organiza informações em **tabelas** (linhas e
colunas), como uma planilha, mas com regras rígidas de tipo e relacionamentos entre tabelas. Cada
tabela tem uma **chave primária** (`PRIMARY KEY`) que identifica cada linha de forma única. O
**SQLite**, embutido no Python via o módulo `sqlite3`, é um banco de dados leve, sem servidor — perfeito
para aprender e para aplicações pequenas/médias.

📌 O fluxo básico: conectar (`sqlite3.connect(...)`), criar um `cursor`, executar comandos SQL
(`cursor.execute(...)`), e confirmar mudanças (`conexao.commit()`). Usamos `:memory:` para criar um
banco temporário na memória, sem gerar arquivos no disco.

**💻 Exemplo comentado:**

````python
import sqlite3

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()

cursor.execute("""
    CREATE TABLE alunos (
        id INTEGER PRIMARY KEY,
        nome TEXT NOT NULL,
        idade INTEGER
    )
""")

cursor.execute("INSERT INTO alunos (nome, idade) VALUES (?, ?)", ("Ana", 25))
cursor.execute("INSERT INTO alunos (nome, idade) VALUES (?, ?)", ("Bruno", 30))
conexao.commit()

cursor.execute("SELECT * FROM alunos")
print(cursor.fetchall())

conexao.close()
````

## Capítulo 2: Consultas Básicas em SQL (SELECT)

`SELECT` é o comando mais usado em SQL: busca dados de uma tabela. `SELECT * FROM
tabela` retorna todas as colunas; `SELECT coluna1, coluna2 FROM tabela` retorna apenas as colunas
escolhidas. `SELECT DISTINCT coluna` remove valores repetidos. Podemos renomear colunas no resultado
com `AS` (ex.: `SELECT nome AS nome_completo`).

**💻 Exemplo comentado:**

````python
import sqlite3

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()
cursor.execute("CREATE TABLE produtos (id INTEGER PRIMARY KEY, nome TEXT, categoria TEXT, preco REAL)")
produtos = [
    ("Mouse", "Periféricos", 89.9), ("Teclado", "Periféricos", 149.9),
    ("Monitor", "Telas", 899.9), ("Webcam", "Periféricos", 199.9),
]
cursor.executemany("INSERT INTO produtos (nome, categoria, preco) VALUES (?, ?, ?)", produtos)
conexao.commit()

cursor.execute("SELECT nome, preco FROM produtos")
print(cursor.fetchall())

cursor.execute("SELECT DISTINCT categoria FROM produtos")
print(cursor.fetchall())

conexao.close()
````

## Capítulo 3: Filtros e Ordenação (WHERE, ORDER BY)

`WHERE` filtra linhas com base em uma condição (`=`, `>`, `<`, `>=`, `<=`, `!=`,
`LIKE` para padrões de texto, `IN` para uma lista de valores, `AND`/`OR` para combinar condições).
`ORDER BY coluna ASC/DESC` ordena o resultado (ASC é o padrão). `LIMIT n` restringe o número de linhas
retornadas — útil para "top N" consultas.

**💻 Exemplo comentado:**

````python
import sqlite3

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()
cursor.execute("CREATE TABLE produtos (id INTEGER PRIMARY KEY, nome TEXT, preco REAL)")
produtos = [("Mouse", 89.9), ("Teclado", 149.9), ("Monitor", 899.9), ("Webcam", 199.9)]
cursor.executemany("INSERT INTO produtos (nome, preco) VALUES (?, ?)", produtos)
conexao.commit()

cursor.execute("SELECT nome, preco FROM produtos WHERE preco > 100 ORDER BY preco DESC")
print(cursor.fetchall())

cursor.execute("SELECT nome FROM produtos WHERE nome LIKE 'M%'")
print(cursor.fetchall())

cursor.execute("SELECT nome, preco FROM produtos ORDER BY preco DESC LIMIT 2")
print(cursor.fetchall())

conexao.close()
````

## Capítulo 4: Agregações (GROUP BY, HAVING)

Funções de agregação resumem várias linhas em um único valor: `COUNT()` (contagem),
`SUM()` (soma), `AVG()` (média), `MIN()`/`MAX()` (mínimo/máximo). `GROUP BY coluna` agrupa linhas que
têm o mesmo valor em uma coluna, aplicando a agregação a cada grupo separadamente. `HAVING` filtra
grupos **depois** da agregação (diferente de `WHERE`, que filtra linhas **antes**).

**💻 Exemplo comentado:**

````python
import sqlite3

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()
cursor.execute("CREATE TABLE vendas (id INTEGER PRIMARY KEY, categoria TEXT, valor REAL)")
vendas = [
    ("Eletrônicos", 500), ("Eletrônicos", 300), ("Livros", 50),
    ("Livros", 30), ("Livros", 20), ("Roupas", 150),
]
cursor.executemany("INSERT INTO vendas (categoria, valor) VALUES (?, ?)", vendas)
conexao.commit()

cursor.execute("SELECT categoria, SUM(valor), COUNT(*) FROM vendas GROUP BY categoria")
print(cursor.fetchall())

cursor.execute("SELECT categoria, SUM(valor) as total FROM vendas GROUP BY categoria HAVING total > 100")
print(cursor.fetchall())

conexao.close()
````

## Capítulo 5: JOINs entre Tabelas

`JOIN` combina linhas de duas tabelas relacionadas por uma chave em comum. O
`INNER JOIN` retorna apenas as linhas que têm correspondência em ambas as tabelas. O `LEFT JOIN`
retorna todas as linhas da tabela da esquerda, mesmo sem correspondência na direita (preenchendo com
`NULL`). JOINs são essenciais porque bancos relacionais separam dados em tabelas diferentes para
evitar repetição (ex.: dados do cliente ficam em uma tabela, pedidos em outra, ligados por
`cliente_id`).

**💻 Exemplo comentado:**

````python
import sqlite3

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()
cursor.execute("CREATE TABLE clientes (id INTEGER PRIMARY KEY, nome TEXT)")
cursor.execute("CREATE TABLE pedidos (id INTEGER PRIMARY KEY, cliente_id INTEGER, valor REAL)")

cursor.executemany("INSERT INTO clientes (id, nome) VALUES (?, ?)", [(1, "Ana"), (2, "Bruno"), (3, "Carla")])
cursor.executemany("INSERT INTO pedidos (cliente_id, valor) VALUES (?, ?)", [(1, 100), (1, 50), (2, 200)])
conexao.commit()

cursor.execute("""
    SELECT clientes.nome, pedidos.valor
    FROM clientes
    INNER JOIN pedidos ON clientes.id = pedidos.cliente_id
""")
print(cursor.fetchall())

# LEFT JOIN inclui Carla mesmo sem pedidos (valor aparece como None)
cursor.execute("""
    SELECT clientes.nome, pedidos.valor
    FROM clientes
    LEFT JOIN pedidos ON clientes.id = pedidos.cliente_id
""")
print(cursor.fetchall())

conexao.close()
````

## Capítulo 6: Subconsultas

Uma **subconsulta** (subquery) é uma consulta SQL dentro de outra consulta. É útil
quando o filtro depende do resultado de outra consulta. Exemplo: "encontre clientes cujo total gasto é
maior que a média geral" exige primeiro calcular a média (uma subconsulta), depois comparar cada
cliente a ela. Subconsultas podem aparecer no `WHERE` (mais comum), no `FROM` (como uma tabela
temporária), ou até no `SELECT`.

**💻 Exemplo comentado:**

````python
import sqlite3

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()
cursor.execute("CREATE TABLE vendas (id INTEGER PRIMARY KEY, vendedor TEXT, valor REAL)")
vendas = [("Ana", 500), ("Bruno", 100), ("Carla", 800), ("Diego", 200)]
cursor.executemany("INSERT INTO vendas (vendedor, valor) VALUES (?, ?)", vendas)
conexao.commit()

# Subconsulta no WHERE: vendedores com valor acima da média geral
cursor.execute("""
    SELECT vendedor, valor FROM vendas
    WHERE valor > (SELECT AVG(valor) FROM vendas)
""")
print(cursor.fetchall())

conexao.close()
````

## Capítulo 7: Conectando Python ao SQL

Python conecta a bancos SQL via módulos como `sqlite3` (nativo). Sempre use
**consultas parametrizadas** (`cursor.execute("... WHERE x = ?", (valor,))`) em vez de concatenar
strings diretamente — isso evita **SQL Injection**, uma vulnerabilidade grave onde um usuário mal-
intencionado injeta comandos SQL através de um campo de entrada. A biblioteca `pandas` também lê
resultados SQL diretamente em um DataFrame com `pd.read_sql_query(sql, conexao)`.

**💻 Exemplo comentado:**

````python
import sqlite3
import pandas as pd

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()
cursor.execute("CREATE TABLE produtos (id INTEGER PRIMARY KEY, nome TEXT, preco REAL)")
cursor.executemany("INSERT INTO produtos (nome, preco) VALUES (?, ?)", [("Mouse", 89.9), ("Teclado", 149.9)])
conexao.commit()

# Consulta parametrizada seguro (evita SQL Injection)
nome_busca = "Mouse"
cursor.execute("SELECT * FROM produtos WHERE nome = ?", (nome_busca,))
print(cursor.fetchall())

# Lendo direto para um DataFrame do pandas
df = pd.read_sql_query("SELECT * FROM produtos", conexao)
print(df)

conexao.close()
````

````python
import sqlite3

conexao = sqlite3.connect(":memory:")
cursor = conexao.cursor()

# 1. Criando as tabelas
cursor.execute("CREATE TABLE clientes (id INTEGER PRIMARY KEY, nome TEXT)")
cursor.execute("CREATE TABLE produtos (id INTEGER PRIMARY KEY, nome TEXT, categoria TEXT, preco REAL)")
cursor.execute("""
    CREATE TABLE pedidos (
        id INTEGER PRIMARY KEY,
        cliente_id INTEGER,
        produto_id INTEGER,
        quantidade INTEGER
    )
""")

# 2. Populando com dados de exemplo
clientes = [(1, "Ana"), (2, "Bruno"), (3, "Carla")]
produtos = [
    (1, "Mouse", "Periféricos", 89.9),
    (2, "Teclado", "Periféricos", 149.9),
    (3, "Monitor", "Telas", 899.9),
]
pedidos = [
    (1, 1, 1, 2),   # Ana comprou 2 mouses
    (2, 1, 3, 1),   # Ana comprou 1 monitor
    (3, 2, 2, 1),   # Bruno comprou 1 teclado
    (4, 3, 1, 3),   # Carla comprou 3 mouses
]

cursor.executemany("INSERT INTO clientes VALUES (?, ?)", clientes)
cursor.executemany("INSERT INTO produtos VALUES (?, ?, ?, ?)", produtos)
cursor.executemany("INSERT INTO pedidos VALUES (?, ?, ?, ?)", pedidos)
conexao.commit()

# 3. Cliente que mais gastou
cursor.execute("""
    SELECT clientes.nome, SUM(produtos.preco * pedidos.quantidade) as total_gasto
    FROM pedidos
    INNER JOIN clientes ON pedidos.cliente_id = clientes.id
    INNER JOIN produtos ON pedidos.produto_id = produtos.id
    GROUP BY clientes.nome
    ORDER BY total_gasto DESC
    LIMIT 1
""")
print("3. Cliente que mais gastou:", cursor.fetchone())

# 4. Produto mais vendido (em unidades)
cursor.execute("""
    SELECT produtos.nome, SUM(pedidos.quantidade) as unidades_vendidas
    FROM pedidos
    INNER JOIN produtos ON pedidos.produto_id = produtos.id
    GROUP BY produtos.nome
    ORDER BY unidades_vendidas DESC
    LIMIT 1
""")
print("4. Produto mais vendido:", cursor.fetchone())

# 5. Receita total por categoria
cursor.execute("""
    SELECT produtos.categoria, SUM(produtos.preco * pedidos.quantidade) as receita
    FROM pedidos
    INNER JOIN produtos ON pedidos.produto_id = produtos.id
    GROUP BY produtos.categoria
    ORDER BY receita DESC
""")
print("5. Receita por categoria:", cursor.fetchall())

conexao.close()
````
