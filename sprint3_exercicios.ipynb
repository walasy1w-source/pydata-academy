## Capítulo 1: Introdução ao NumPy

Listas Python são flexíveis, mas lentas para cálculos numéricos em grande escala.
O **NumPy** resolve isso com o `array` — uma estrutura homogênea (todos os itens do mesmo tipo) que
permite **operações vetorizadas**: em vez de um `for` para multiplicar cada item por 2, escrevemos
`array * 2` e o NumPy faz isso internamente, em código otimizado, muito mais rápido.

📌 Também é possível **filtrar com condições booleanas**: `array[array > 5]` retorna só os itens
maiores que 5, sem escrever um loop.

**💻 Exemplo comentado:**

````python
import numpy as np

notas = np.array([8, 7.5, 9, 6, 8.5])

print(notas.mean())        # média
print(notas.std())         # desvio padrão
print(notas * 10)          # operação vetorizada (multiplica cada item)
print(notas[notas > 7])    # filtragem booleana

matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(matriz.shape)        # (linhas, colunas)
````

## Capítulo 3: Leitura e Escrita de Arquivos (CSV, Excel, JSON)

Dados raramente nascem dentro do seu código — eles vêm de arquivos. pandas lê e
grava os formatos mais comuns com uma linha: `pd.read_csv()`/`.to_csv()` para CSV,
`pd.read_excel()`/`.to_excel()` para Excel (requer a biblioteca `openpyxl`), e
`pd.read_json()`/`.to_json()` para JSON.

**💻 Exemplo comentado:**

````python
import pandas as pd

df = pd.DataFrame({"produto": ["Caneta", "Caderno"], "preco": [2.5, 15.0]})

df.to_csv("produtos_temp.csv", index=False)               # grava CSV
df_lido = pd.read_csv("produtos_temp.csv")                 # lê CSV
print(df_lido)

df.to_json("produtos_temp.json", orient="records")         # grava JSON
df_json = pd.read_json("produtos_temp.json")                # lê JSON
print(df_json)

import os
os.remove("produtos_temp.csv")
os.remove("produtos_temp.json")
````

## Capítulo 4: Limpeza de Dados

Dados reais quase sempre têm problemas: valores ausentes (`NaN`), linhas duplicadas
ou colunas com o tipo errado. pandas oferece `.isna()`/`.dropna()`/`.fillna()` para lidar com valores
ausentes, `.duplicated()`/`.drop_duplicates()` para duplicatas, e `.astype()` para converter tipos.

📌 A decisão entre **remover** (`dropna`) ou **preencher** (`fillna`) um valor ausente depende do
contexto — remover perde dados, preencher pode introduzir viés. Não existe resposta única certa.

**💻 Exemplo comentado:**

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "produto": ["Caneta", "Caderno", "Caneta", None],
    "preco": [2.5, np.nan, 2.5, 10.0],
})

print(df.isna())                          # onde há valores ausentes
print(df.dropna())                        # remove linhas com qualquer NaN
print(df.fillna({"preco": 0, "produto": "Desconhecido"}))  # preenche valores ausentes
print(df.drop_duplicates())                # remove linhas duplicadas
````

## Capítulo 5: Filtragem e Agrupamento de Dados

Assim como em NumPy, DataFrames aceitam **filtragem booleana**:
`df[df["preco"] > 100]` retorna só as linhas onde a condição é verdadeira. Para resumir dados por
categoria, usamos `.groupby()` — por exemplo, "total vendido por produto" — combinado com uma função
de agregação como `.sum()`, `.mean()` ou `.agg()` (para várias agregações de uma vez).

**💻 Exemplo comentado:**

````python
import pandas as pd

df = pd.DataFrame({
    "produto": ["Caneta", "Caderno", "Caneta", "Mochila"],
    "quantidade": [10, 5, 6, 2],
    "preco": [2.5, 15.0, 2.5, 120.0],
})

caros = df[df["preco"] > 10]                          # filtragem booleana
print(caros)

total_por_produto = df.groupby("produto")["quantidade"].sum()   # agrupamento simples
print(total_por_produto)

resumo = df.groupby("produto").agg({"quantidade": "sum", "preco": "mean"})
print(resumo)
````

## Capítulo 6: Junção de Tabelas (merge, join, concat)

Dados do mundo real geralmente estão espalhados em várias tabelas — clientes em uma,
pedidos em outra. `pd.merge()` combina DataFrames por uma coluna em comum (a "chave"), de forma
parecida com um JOIN em SQL (`how="inner"` mantém só o que existe nos dois lados; `how="left"`
mantém tudo do DataFrame da esquerda, preenchendo com `NaN` o que não encontrar). Já `pd.concat()`
simplesmente empilha DataFrames, um em cima do outro (ou lado a lado).

**💻 Exemplo comentado:**

````python
import pandas as pd

clientes = pd.DataFrame({"cliente_id": [1, 2, 3], "nome": ["Ana", "Bruno", "Carla"]})
pedidos = pd.DataFrame({"cliente_id": [1, 2, 2], "valor": [100, 200, 50]})

# merge tipo "inner": só clientes que têm pedido
combinado = pd.merge(clientes, pedidos, on="cliente_id", how="inner")
print(combinado)

# merge tipo "left": todos os clientes, mesmo sem pedido (vira NaN)
combinado_left = pd.merge(clientes, pedidos, on="cliente_id", how="left")
print(combinado_left)

# concat: empilha DataFrames
novos_clientes = pd.DataFrame({"cliente_id": [4], "nome": ["Diego"]})
todos = pd.concat([clientes, novos_clientes], ignore_index=True)
print(todos)
````

## Capítulo 7: Visualização de Dados com Matplotlib

Um bom gráfico comunica em segundos o que uma tabela levaria minutos para explicar.
`matplotlib.pyplot` (convencionalmente importado como `plt`) é a biblioteca de visualização mais
usada em Python: `plt.plot()` para linhas, `plt.bar()` para barras, `plt.pie()` para pizza. Sempre
vale adicionar `título`, `xlabel`/`ylabel` e, se houver mais de uma série, uma `legend()`.

**💻 Exemplo comentado:**

````python
import matplotlib
matplotlib.use("Agg")  # backend sem tela, só para este notebook rodar em qualquer ambiente
import matplotlib.pyplot as plt

meses = ["Jan", "Fev", "Mar", "Abr"]
vendas = [1200, 1500, 900, 1800]

plt.figure(figsize=(6, 4))
plt.bar(meses, vendas, color="#d4af37")
plt.title("Vendas por mês")
plt.xlabel("Mês")
plt.ylabel("Vendas (R$)")
plt.tight_layout()
plt.show()
print("✅ Gráfico gerado com sucesso!")
````

### ✅ Gabarito do projeto

````python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

np.random.seed(42)
categorias = ["Eletrônicos", "Papelaria", "Vestuário", "Casa"]
n = 40
dataset = pd.DataFrame({
    "pedido_id": range(1, n + 1),
    "categoria": np.random.choice(categorias, n),
    "quantidade": np.random.randint(1, 5, n),
    "preco_unitario": np.round(np.random.uniform(10, 300, n), 2),
})
dataset.loc[3, "preco_unitario"] = np.nan
dataset.loc[10, "categoria"] = None
dataset = pd.concat([dataset, dataset.iloc[[5]]], ignore_index=True)

# 1. Limpeza: remove duplicatas e trata valores ausentes
dataset = dataset.drop_duplicates(subset="pedido_id")
dataset["preco_unitario"] = dataset["preco_unitario"].fillna(dataset["preco_unitario"].mean())
dataset["categoria"] = dataset["categoria"].fillna("Não informado")

# 2. Faturamento por pedido e total
dataset["faturamento"] = dataset["quantidade"] * dataset["preco_unitario"]
faturamento_total = dataset["faturamento"].sum()
print(f"Faturamento total: R$ {faturamento_total:.2f}")

# 3. Faturamento por categoria
faturamento_por_categoria = dataset.groupby("categoria")["faturamento"].sum().sort_values(ascending=False)
print(faturamento_por_categoria)

# 4. Categoria com maior faturamento
categoria_top = faturamento_por_categoria.idxmax()
print(f"Categoria com maior faturamento: {categoria_top}")

# 5. Visualização
plt.figure(figsize=(7, 4))
plt.bar(faturamento_por_categoria.index, faturamento_por_categoria.values, color="#d4af37")
plt.title("Faturamento por categoria")
plt.ylabel("Faturamento (R$)")
plt.tight_layout()
plt.show()
print("✅ Análise concluída!")
````
