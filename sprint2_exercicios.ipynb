## Capítulo 1: Compreensão de Listas (List Comprehensions)

Uma *list comprehension* é uma forma compacta de criar uma lista a partir de um
loop, tudo em uma linha só. Em vez de escrever um `for` de 3 linhas para preencher uma lista vazia,
escrevemos `[expressao for item in iteravel]` — e ainda podemos adicionar uma condição no final:
`[expressao for item in iteravel if condicao]`.

📌 Isso não é só "mais bonito": código com list comprehension costuma ser mais rápido e mais fácil
de ler, uma vez que você se acostuma com a sintaxe.

**💻 Exemplo comentado:**

````python
numeros = [1, 2, 3, 4, 5]

quadrados = [n ** 2 for n in numeros]          # aplica uma expressão a cada item
pares = [n for n in numeros if n % 2 == 0]     # filtra com "if"

print(quadrados)
print(pares)
````

## Capítulo 2: Manipulação Avançada de Strings

Além dos métodos básicos, f-strings aceitam **especificadores de formatação**:
`f"{valor:.2f}"` mostra 2 casas decimais, `f"{valor:>10}"` alinha à direita em um campo de 10

**💻 Exemplo comentado:**

````python
preco = 49.9
print(f"{preco:>10.2f}")  # alinhado à direita, 2 casas decimais, campo de 10 caracteres

nome = "  python  "
print(nome.strip().upper().replace("PYTHON", "Python Pro"))  # métodos encadeados

email = "aluno@pydata.academy"
print(email.endswith("@pydata.academy"))
````

## Capítulo 3: Tratamento de Exceções (try/except)

Nem todo código dá certo na primeira tentativa — tentar converter `"abc"` para
número, dividir por zero, ou acessar uma chave que não existe em um dicionário são erros comuns.
Em vez de deixar o programa quebrar, usamos `try/except` para tratar esses erros com elegância.

📌 Boas práticas: capture o tipo **específico** de exceção (`ValueError`, `ZeroDivisionError`,
`KeyError`) em vez de um `except:` genérico — assim você não esconde bugs inesperados. O bloco
`else` roda se **não** houve erro, e `finally` roda **sempre**, com ou sem erro.

**💻 Exemplo comentado:**

````python
try:
    numero = int("abc")
except ValueError as e:
    print(f"Erro: {e}")
else:
    print("Conversão OK")
finally:
    print("Tentativa concluída")
````

## Capítulo 5: Trabalhando com Arquivos

Para ler ou gravar arquivos, usamos `open()` — sempre de preferência com `with`,
que fecha o arquivo automaticamente ao final do bloco, mesmo se ocorrer um erro. Os modos mais
comuns são `"r"` (leitura), `"w"` (escrita, sobrescreve o arquivo) e `"a"` (append, adiciona ao
final sem apagar o que já existia).

**💻 Exemplo comentado:**

````python
with open("exemplo_temp.txt", "w", encoding="utf-8") as f:
    f.write("Python\n")
    f.write("Ciência de Dados\n")

with open("exemplo_temp.txt", "r", encoding="utf-8") as f:
    conteudo = f.read()
print(conteudo)

import os
os.remove("exemplo_temp.txt")  # limpeza: remove o arquivo de exemplo
````

## Capítulo 6: Programação Orientada a Objetos I

Uma **classe** é um molde para criar objetos que guardam dados (atributos) e
comportamentos (métodos) juntos. `self` representa a própria instância dentro da classe — é como
o objeto se refere a si mesmo. O método especial `__init__` roda automaticamente quando um objeto
é criado, e serve para inicializar seus atributos.

**💻 Exemplo comentado:**

````python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

    def aplicar_desconto(self, percentual):
        self.preco = self.preco * (1 - percentual)

produto1 = Produto("Notebook", 3200)
produto1.aplicar_desconto(0.1)
print(produto1.nome, produto1.preco)
````

**Enunciado:** construa um mini sistema de estoque para a **PyData Store**, combinando tudo que foi

**Requisitos:**

2. Uma classe `Estoque` que guarda uma lista de produtos e oferece:
   - `adicionar(produto)`
   - `produtos_em_falta(limite=5)` — lista (via list comprehension) os nomes dos produtos com
     quantidade abaixo do limite.
   - `valor_total_estoque()` — soma o valor total de todos os produtos.
   - `buscar(nome)` — retorna o produto pelo nome, ou levanta `KeyError` se não existir.
3. Trate o `KeyError` ao buscar um produto inexistente, imprimindo um aviso amigável.
4. Grave um relatório simples (`nome;quantidade;preco`) em um arquivo de texto, releia e imprima.

### ✅ Gabarito do projeto

````python
class Produto:
    def __init__(self, nome, preco, quantidade):
        self.nome = nome
        self.preco = preco
        self.quantidade = quantidade

    def valor_total(self):
        return self.preco * self.quantidade


class Estoque:
    def __init__(self):
        self.produtos = []

    def adicionar(self, produto):
        self.produtos.append(produto)

    def produtos_em_falta(self, limite=5):
        return [p.nome for p in self.produtos if p.quantidade < limite]

    def valor_total_estoque(self):
        return sum(p.valor_total() for p in self.produtos)

    def buscar(self, nome):
        try:
            return next(p for p in self.produtos if p.nome.lower() == nome.lower())
        except StopIteration:
            raise KeyError(f"Produto \'{nome}\' não encontrado")


estoque = Estoque()
estoque.adicionar(Produto("Caneta", 2.5, 100))
estoque.adicionar(Produto("Caderno", 15.0, 3))
estoque.adicionar(Produto("Mochila", 120.0, 2))

print(f"Valor total do estoque: R$ {estoque.valor_total_estoque():.2f}")
print(f"Produtos em falta (menos de 5 unidades): {estoque.produtos_em_falta()}")

try:
    estoque.buscar("Tablet")
except KeyError as e:
    print(f"Aviso: {e}")

with open("relatorio_estoque.txt", "w", encoding="utf-8") as f:
    for p in estoque.produtos:
        f.write(f"{p.nome};{p.quantidade};{p.preco:.2f}\n")

with open("relatorio_estoque.txt", "r", encoding="utf-8") as f:
    print(f.read())

import os
os.remove("relatorio_estoque.txt")  # limpeza
````
