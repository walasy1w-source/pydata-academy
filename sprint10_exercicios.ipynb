Antes de treinar qualquer modelo, é preciso traduzir uma pergunta de negócio vaga
("como reduzir cancelamentos?") em um **problema de ML bem definido**: qual é a **variável-alvo** (o
que queremos prever), quais **features** (dados disponíveis) usar, e que **tipo de problema** é esse
(classificação, regressão, ou clustering). Um erro comum é pular direto para o algoritmo sem antes
fazer essa tradução — o que quase sempre leva a um modelo tecnicamente correto, mas inútil na
prática.

**💻 Exemplo comentado:**

## Capítulo 2: Segmentação de Clientes

**Segmentação de clientes** agrupa clientes parecidos entre si (por comportamento de
compra, engajamento, etc.) usando clustering (Sprint 8/9), sem que ninguém precise definir os grupos
manualmente. Um framework clássico é o **RFM**: Recência (há quanto tempo comprou pela última vez),
Frequência (quantas vezes comprou) e Valor Monetário (quanto gastou no total). Depois de agrupar,
interpretamos cada cluster (ex.: "clientes VIP", "clientes em risco de abandono") para guiar ações de
marketing diferentes por grupo.

**💻 Exemplo comentado:**

````python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

clientes = pd.DataFrame({
    "recencia_dias": [5, 10, 200, 180, 3, 220],
    "frequencia_compras": [20, 18, 2, 3, 25, 1],
    "valor_total": [5000, 4500, 150, 200, 6000, 100],
})

modelo = KMeans(n_clusters=2, random_state=0, n_init=10)
clientes["cluster"] = modelo.fit_predict(clientes)

print(clientes)
print("\nCentros dos clusters:")
print(pd.DataFrame(modelo.cluster_centers_, columns=clientes.columns[:-1]))
````

## Capítulo 3: Previsão de Churn

**Churn** é a taxa de clientes que abandonam um serviço. Prever churn é um problema
clássico de classificação: usamos dados históricos de comportamento (uso, reclamações, tempo de
contrato) como features, e "cancelou" (0/1) como alvo. O valor de negócio está em **agir antes** do
cancelamento — por isso o **recall** (encontrar o máximo de clientes em risco) costuma importar mais
que a precisão pura: é mais barato oferecer um desconto a alguém que não ia cancelar do que perder um
cliente que não foi identificado a tempo.

**💻 Exemplo comentado:**

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np

np.random.seed(42)
n = 300
tempo_contrato_meses = np.random.randint(1, 48, n)
reclamacoes = np.random.poisson(0.7, n)
uso_mensal_horas = np.random.normal(15, 6, n)

probabilidade_churn = 1 / (1 + np.exp(-(reclamacoes * 1.1 - uso_mensal_horas * 0.1 - tempo_contrato_meses * 0.02)))
churn = (probabilidade_churn > np.random.rand(n)).astype(int)

X = np.column_stack([tempo_contrato_meses, reclamacoes, uso_mensal_horas])
y = churn

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)
modelo = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
modelo.fit(X_treino, y_treino)

print(classification_report(y_teste, modelo.predict(X_teste)))
````

## Capítulo 4: Análise de Risco de Crédito

Modelos de **risco de crédito** preveem a probabilidade de um cliente não pagar um
empréstimo/fatura ("inadimplência" ou "default"). Aqui, o **limiar de decisão** (threshold) importa
muito: por padrão, muitos modelos usam 0.5 de probabilidade para decidir, mas em crédito, às vezes
vale a pena ser mais conservador (limiar mais baixo, recusando mais pedidos) para reduzir o risco de
grandes perdas financeiras — uma decisão de negócio, não só técnica.

**💻 Exemplo comentado:**

````python
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([[30000, 2], [80000, 0], [15000, 4], [90000, 1], [20000, 3], [100000, 0]])  # renda, dívidas ativas
y = np.array([1, 0, 1, 0, 1, 0])  # 1 = inadimplente, 0 = bom pagador

modelo = LogisticRegression()
modelo.fit(X, y)

cliente_novo = [[25000, 3]]
probabilidade_inadimplencia = modelo.predict_proba(cliente_novo)[0][1]
print(f"Probabilidade de inadimplência: {probabilidade_inadimplencia:.2f}")

# Testando limiares diferentes
for limiar in [0.5, 0.3, 0.2]:
    decisao = "RECUSAR" if probabilidade_inadimplencia >= limiar else "APROVAR"
    print(f"Limiar {limiar}: {decisao}")
````

## Capítulo 5: Otimização de Receita

ML pode otimizar receita além de só prever: uma estratégia comum é calcular o
**valor esperado** de uma ação — `probabilidade × valor` — para priorizar onde investir esforço
limitado (ex.: um time de retenção não consegue ligar para todos os clientes em risco, então liga
primeiro para quem tem maior `probabilidade_churn × valor_do_cliente`, maximizando o impacto do
esforço).

**💻 Exemplo comentado:**

````python
import pandas as pd

clientes = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carla", "Diego"],
    "probabilidade_churn": [0.8, 0.3, 0.6, 0.9],
    "valor_mensal": [50, 500, 100, 30],
})

clientes["valor_esperado_perda"] = clientes["probabilidade_churn"] * clientes["valor_mensal"]
clientes_priorizados = clientes.sort_values("valor_esperado_perda", ascending=False)

print(clientes_priorizados)
````

## Capítulo 6: Comunicação de Resultados para Stakeholders

Um modelo tecnicamente excelente que ninguém entende não gera impacto. Ao comunicar
resultados de ML para **stakeholders não técnicos**, traduza métricas em linguagem de negócio: em vez
de "o modelo tem 85% de acurácia", diga "o modelo identifica corretamente 85 em cada 100 clientes";
em vez de "recall de 0.9", diga "o modelo encontra 9 em cada 10 clientes que realmente cancelariam".
Sempre conecte o resultado técnico a um **impacto financeiro ou operacional** concreto.

**💻 Exemplo comentado:**

````python
resultado_tecnico = {
    "acuracia": 0.85,
    "recall": 0.9,
    "clientes_identificados_corretamente": 90,
    "total_clientes_risco_real": 100,
    "valor_medio_cliente_retido": 150,
}

resumo_executivo = f"""
RESUMO EXECUTIVO — Modelo de Previsão de Churn

O modelo identifica corretamente {resultado_tecnico['acuracia']*100:.0f}% dos casos no geral.
Mais importante: de cada {resultado_tecnico['total_clientes_risco_real']} clientes que realmente
cancelariam, o modelo encontra {resultado_tecnico['clientes_identificados_corretamente']} — ou seja,
{resultado_tecnico['recall']*100:.0f}% dos casos de risco real.

Se a equipe de retenção agir sobre esses {resultado_tecnico['clientes_identificados_corretamente']}
clientes identificados, e conseguir reter metade deles (estimativa conservadora), o impacto seria de
aproximadamente R$ {resultado_tecnico['clientes_identificados_corretamente'] * 0.5 * resultado_tecnico['valor_medio_cliente_retido']:.2f} em receita mensal preservada.
"""
print(resumo_executivo)
````

````python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score

# Dataset sintético de assinantes
np.random.seed(7)
n = 200
dados = pd.DataFrame({
    "recencia_dias": np.random.randint(1, 300, n),
    "frequencia_compras": np.random.randint(1, 30, n),
    "valor_mensal": np.random.normal(100, 40, n).round(2),
    "reclamacoes": np.random.poisson(0.7, n),
})
probabilidade_churn = 1 / (1 + np.exp(-(dados["reclamacoes"] * 1.1 + dados["recencia_dias"] * 0.01 - dados["frequencia_compras"] * 0.1)))
dados["cancelou"] = (probabilidade_churn > np.random.rand(n)).astype(int)

# 1. Segmentação de clientes
modelo_cluster = KMeans(n_clusters=3, random_state=7, n_init=10)
dados["cluster"] = modelo_cluster.fit_predict(dados[["recencia_dias", "frequencia_compras", "valor_mensal"]])
print("1. Segmentação:")
print(dados.groupby("cluster")["valor_mensal"].mean())

# 2. Modelo de previsão de churn
X = dados[["recencia_dias", "frequencia_compras", "reclamacoes"]]
y = dados["cancelou"]
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.25, random_state=7)
modelo_churn = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=7)
modelo_churn.fit(X_treino, y_treino)
recall_teste = recall_score(y_teste, modelo_churn.predict(X_teste))
print(f"\n2. Recall no teste: {recall_teste:.3f}")

# 3. Valor esperado e priorização
dados["probabilidade_churn"] = modelo_churn.predict_proba(X)[:, 1]
dados["valor_esperado_perda"] = dados["probabilidade_churn"] * dados["valor_mensal"]
top10 = dados.sort_values("valor_esperado_perda", ascending=False).head(10)
valor_esperado_top10 = top10["valor_esperado_perda"].sum()
print(f"\n3. Top 10 clientes prioritários (valor esperado total: R$ {valor_esperado_top10:.2f})")

# 4. Resumo executivo
impacto_estimado = top10["valor_esperado_perda"].sum() * 0.5  # supondo 50% de retenção
resumo_executivo = f"""
RESUMO EXECUTIVO — Caso de Negócio: Retenção de Assinantes (PyData Streaming)

Identificamos 3 segmentos de clientes por comportamento de compra. O modelo de previsão de churn
encontra {recall_teste*100:.0f}% dos clientes que realmente cancelariam (recall no teste).

Priorizando os 10 clientes com maior valor esperado de perda, e assumindo uma taxa de retenção de
50% nas ações da equipe, o impacto estimado é de R$ {impacto_estimado:.2f} em receita mensal
preservada.
"""
print("\n4. Resumo executivo:")
print(resumo_executivo)
````
