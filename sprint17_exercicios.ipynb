## Capítulo 2: Desenvolvimento e Modelagem

Com o plano definido, a fase de desenvolvimento integra as técnicas dos sprints
anteriores: coleta/geração de dados, limpeza (Sprint 3), análise exploratória (Sprint 3/4), e
modelagem (Sprint 8/9/16, dependendo do tipo de problema). O exemplo abaixo integra **classificação
supervisionada** (prever churn) e **clustering** (segmentar clientes) no mesmo dataset — mostrando como
combinar mais de uma técnica no mesmo projeto, algo comum em projetos de portfólio robustos.

**💻 Exemplo guiado:**

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score
from sklearn.cluster import KMeans

# Geração e limpeza de dados (Sprint 3)
np.random.seed(42)
n = 300
dados = pd.DataFrame({
    "meses_assinatura": np.random.randint(1, 48, n),
    "horas_assistidas_mes": np.random.normal(20, 8, n).round(1),
    "reclamacoes_suporte": np.random.poisson(0.7, n),
})
dados.loc[3, "horas_assistidas_mes"] = np.nan   # simula dado ausente real
dados["horas_assistidas_mes"] = dados["horas_assistidas_mes"].fillna(dados["horas_assistidas_mes"].mean())

# Variável-alvo sintética (churn)
probabilidade_churn = 1 / (1 + np.exp(-(dados["reclamacoes_suporte"] * 1.1 - dados["horas_assistidas_mes"] * 0.08)))
dados["cancelou"] = (probabilidade_churn > np.random.rand(n)).astype(int)

# Modelagem supervisionada: previsão de churn (Sprint 9)
X = dados[["meses_assinatura", "horas_assistidas_mes", "reclamacoes_suporte"]]
y = dados["cancelou"]
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.25, random_state=42)
modelo_churn = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
modelo_churn.fit(X_treino, y_treino)
recall = recall_score(y_teste, modelo_churn.predict(X_teste))
print(f"Recall do modelo de churn: {recall:.3f}")

# Modelagem não supervisionada: segmentação (Sprint 16)
modelo_cluster = KMeans(n_clusters=3, random_state=42, n_init=10)
dados["segmento"] = modelo_cluster.fit_predict(X)
print("\nPerfil médio por segmento:")
print(dados.groupby("segmento")[["meses_assinatura", "horas_assistidas_mes", "reclamacoes_suporte"]].mean().round(2))
````

## 🏆 Entrega Final — Seu Projeto de Portfólio

Esta célula final junta as 3 etapas (planejamento, desenvolvimento e apresentação) em um exemplo
completo — o modelo do que sua entrega final de portfólio deve conter. Substitua os dados e o
problema pelos do SEU projeto escolhido.


````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score, accuracy_score
from sklearn.cluster import KMeans

# ETAPA 1: Planejamento
canvas_projeto = {
    "titulo": "Previsão de Churn e Segmentação de Assinantes - PyData Streaming",
    "pergunta_de_negocio": "Quais assinantes tem maior risco de cancelar, e quais segmentos existem na base?",
    "criterio_de_sucesso": "Recall >= 0.7 na previsao de churn + segmentos interpretaveis",
}
titulo_projeto = canvas_projeto["titulo"]
print("ETAPA 1 - Planejamento:")
print(f"  {titulo_projeto}\n")

# ETAPA 2: Desenvolvimento e modelagem
np.random.seed(42)
n = 300
dados = pd.DataFrame({
    "meses_assinatura": np.random.randint(1, 48, n),
    "horas_assistidas_mes": np.random.normal(20, 8, n).round(1),
    "reclamacoes_suporte": np.random.poisson(0.7, n),
})
dados.loc[3, "horas_assistidas_mes"] = np.nan
dados["horas_assistidas_mes"] = dados["horas_assistidas_mes"].fillna(dados["horas_assistidas_mes"].mean())

probabilidade_churn = 1 / (1 + np.exp(-(dados["reclamacoes_suporte"] * 1.1 - dados["horas_assistidas_mes"] * 0.08)))
dados["cancelou"] = (probabilidade_churn > np.random.rand(n)).astype(int)

X = dados[["meses_assinatura", "horas_assistidas_mes", "reclamacoes_suporte"]]
y = dados["cancelou"]
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.25, random_state=42)
modelo_churn = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
modelo_churn.fit(X_treino, y_treino)
recall = recall_score(y_teste, modelo_churn.predict(X_teste))

modelo_cluster = KMeans(n_clusters=3, random_state=42, n_init=10)
dados["segmento"] = modelo_cluster.fit_predict(X)

segmentos_encontrados = sorted(dados["segmento"].unique())
print(f"ETAPA 2 - Modelo treinado. Recall no teste: {recall:.3f}")
print(f"  Segmentos encontrados: {segmentos_encontrados}\n")

# ETAPA 3: Apresentação e portfólio
readme_projeto = """# Previsao de Churn e Segmentacao de Assinantes

## O que e
Projeto que preve risco de cancelamento e segmenta assinantes por comportamento de uso.

## Como rodar
pip install numpy pandas scikit-learn

## Principais resultados
Recall de {:.0%} na previsao de churn, com 3 segmentos de cliente identificados.
""".format(recall)

resumo_executivo = f"""
RESUMO EXECUTIVO

O modelo identifica {recall*100:.0f}% dos assinantes que realmente cancelariam, permitindo acoes
de retencao direcionadas. A segmentacao revelou perfis de uso distintos entre os clientes.
"""

print("ETAPA 3 - Documentação gerada:")
print(readme_projeto)
print(resumo_executivo)

print("Projeto Final de Portfolio concluido de ponta a ponta!")
````
