## Capítulo 1: O que é Machine Learning?

Programação tradicional: você escreve **regras explícitas** (`if idade >= 18: maior =
True`) e o computador aplica essas regras aos dados. **Machine Learning (ML)** inverte essa lógica: em
vez de escrever as regras, você mostra **exemplos** (dados + respostas corretas) e o algoritmo
**aprende** os padrões sozinho, gerando um modelo capaz de prever respostas para dados novos, nunca
vistos antes.

📌 ML brilha quando as regras são complexas demais para escrever manualmente (reconhecer rostos,
prever preços de imóveis, detectar fraude) — situações onde é mais fácil mostrar exemplos do que
codificar todas as exceções.

**💻 Exemplo comentado:**

````python
# Programação tradicional: regra explícita escrita à mão
def eh_maior_idade_regra(idade):
    return idade >= 18

print("Regra manual:", eh_maior_idade_regra(20), eh_maior_idade_regra(15))

# "Aprendendo" um limiar a partir de exemplos (uma versão simplificada de ML)
exemplos = [(16, False), (17, False), (18, True), (25, True), (15, False), (30, True)]

# O "modelo" aprende o menor valor rotulado como True
limiar_aprendido = min(idade for idade, rotulo in exemplos if rotulo)
print(f"Limiar aprendido a partir dos dados: {limiar_aprendido}")

def eh_maior_idade_aprendido(idade):
    return idade >= limiar_aprendido

print("Modelo aprendido:", eh_maior_idade_aprendido(20), eh_maior_idade_aprendido(15))
````

## Capítulo 2: Tipos de Aprendizado

Os três grandes tipos: **Aprendizado Supervisionado** (temos dados **rotulados** —
sabemos a resposta certa para cada exemplo de treino; ex.: prever preço de imóvel, classificar e-mail
como spam); **Aprendizado Não Supervisionado** (não há rótulos — o algoritmo encontra estrutura
sozinha, como agrupar clientes parecidos, sem saber de antemão os grupos); **Aprendizado por Reforço**
(um agente aprende por tentativa e erro, recebendo recompensas — usado em jogos e robótica, fora do
escopo deste curso).

Dentro do supervisionado, dividimos em **classificação** (prever uma categoria: spam/não-spam) e
**regressão** (prever um número contínuo: preço, temperatura).

**💻 Exemplo comentado:**

````python
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
import numpy as np

# Supervisionado - regressão: prevendo um valor numérico a partir de dados rotulados (X, y)
X_treino = np.array([[1], [2], [3], [4], [5]])   # horas estudadas
y_treino = np.array([50, 60, 70, 80, 90])         # nota (rótulo conhecido)

modelo_regressao = LinearRegression()
modelo_regressao.fit(X_treino, y_treino)
print("Previsão para 6 horas de estudo:", modelo_regressao.predict([[6]]))

# Não supervisionado - clustering: agrupando dados SEM rótulos
X_sem_rotulo = np.array([[1, 1], [1.5, 2], [10, 10], [10.5, 9.5]])
modelo_cluster = KMeans(n_clusters=2, random_state=0, n_init=10)
grupos = modelo_cluster.fit_predict(X_sem_rotulo)
print("Grupos encontrados (sem rótulos prévios):", grupos)
````

## Capítulo 3: Preparação de Dados para Modelos

Modelos de ML esperam dados **numéricos** e, muitas vezes, em **escalas
comparáveis**. Duas etapas comuns: **normalização/padronização** (`StandardScaler` transforma cada
coluna para ter média 0 e desvio 1 — evita que uma coluna com valores grandes, como salário, domine
uma coluna com valores pequenos, como idade) e **codificação de categorias** (`pd.get_dummies` ou
`OneHotEncoder` transforma texto como "cor: azul/verde/vermelho" em colunas numéricas 0/1). Também
separamos os dados em **treino** e **teste** (`train_test_split`) para avaliar o modelo em dados que
ele nunca viu.

**💻 Exemplo comentado:**

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

df = pd.DataFrame({
    "idade": [25, 40, 22, 35, 50],
    "salario": [3000, 8000, 2500, 6000, 12000],
    "cidade": ["SP", "RJ", "SP", "MG", "RJ"],
})

# Codificação de categoria (one-hot encoding)
df_codificado = pd.get_dummies(df, columns=["cidade"])
print(df_codificado)

# Normalização das colunas numéricas
scaler = StandardScaler()
df_normalizado = df_codificado.copy()
df_normalizado[["idade", "salario"]] = scaler.fit_transform(df[["idade", "salario"]])
print(df_normalizado[["idade", "salario"]])

# Separando treino e teste
X = df_normalizado.drop(columns=[])
X_treino, X_teste = train_test_split(X, test_size=0.2, random_state=42)
print(f"Treino: {len(X_treino)} linhas, Teste: {len(X_teste)} linhas")
````

## Capítulo 4: Métricas de Avaliação

Para **classificação**: `accuracy` (% de acertos totais), `precision` (dos que o
modelo previu como positivos, quantos realmente eram), `recall` (dos que realmente eram positivos,
quantos o modelo encontrou), `f1_score` (média harmônica entre precision e recall). Para
**regressão**: `MAE` (erro médio absoluto), `MSE` (erro médio quadrático, penaliza mais erros
grandes), `R²` (o quanto o modelo explica a variação dos dados, de 0 a 1 — quanto mais perto de 1,
melhor).

**💻 Exemplo comentado:**

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Classificação
y_real = [1, 0, 1, 1, 0, 1]
y_previsto = [1, 0, 0, 1, 0, 1]

print("Accuracy:", accuracy_score(y_real, y_previsto))
print("Precision:", precision_score(y_real, y_previsto))
print("Recall:", recall_score(y_real, y_previsto))
print("F1:", f1_score(y_real, y_previsto))

# Regressão
y_real_reg = [100, 150, 200, 250]
y_previsto_reg = [110, 140, 210, 230]

print("\nMAE:", mean_absolute_error(y_real_reg, y_previsto_reg))
print("MSE:", mean_squared_error(y_real_reg, y_previsto_reg))
print("R²:", r2_score(y_real_reg, y_previsto_reg))
````

## Capítulo 5: Overfitting e Underfitting

**Overfitting** (sobreajuste): o modelo "decora" os dados de treino, incluindo seu
ruído, e por isso vai muito bem no treino mas mal em dados novos (teste) — não generaliza.
**Underfitting** (subajuste): o modelo é simples demais para capturar até os padrões básicos, indo mal
tanto no treino quanto no teste. O ideal é um meio-termo: bom desempenho em treino **e** teste. Uma
forma clássica de detectar overfitting é comparar a métrica de avaliação no treino vs. no teste — uma
diferença grande (treino muito melhor que teste) é sinal de alerta.

**💻 Exemplo comentado:**

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

np.random.seed(0)
X = np.random.rand(100, 2)
y = (X[:, 0] + X[:, 1] > 1).astype(int)

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

# Árvore muito profunda (propensa a overfitting)
modelo_profundo = DecisionTreeClassifier(max_depth=None, random_state=42)
modelo_profundo.fit(X_treino, y_treino)
acc_treino = accuracy_score(y_treino, modelo_profundo.predict(X_treino))
acc_teste = accuracy_score(y_teste, modelo_profundo.predict(X_teste))
print(f"Árvore profunda -> treino: {acc_treino:.2f}, teste: {acc_teste:.2f}")

# Árvore rasa (mais controlada)
modelo_raso = DecisionTreeClassifier(max_depth=2, random_state=42)
modelo_raso.fit(X_treino, y_treino)
acc_treino_raso = accuracy_score(y_treino, modelo_raso.predict(X_treino))
acc_teste_raso = accuracy_score(y_teste, modelo_raso.predict(X_teste))
print(f"Árvore rasa -> treino: {acc_treino_raso:.2f}, teste: {acc_teste_raso:.2f}")
````

## Capítulo 6: Introdução ao Scikit-learn

O `scikit-learn` é a biblioteca de ML mais usada em Python, com uma API consistente
para todos os modelos: `modelo.fit(X, y)` treina, `modelo.predict(X_novo)` prevê, e (para alguns
modelos) `modelo.predict_proba(X_novo)` retorna probabilidades. Essa padronização significa que trocar
de um algoritmo para outro (ex.: de árvore de decisão para regressão logística) muitas vezes é só
trocar o nome da classe — o resto do código continua igual.

**💻 Exemplo comentado:**

````python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
import numpy as np

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

# Mesma API, dois algoritmos diferentes
for ModeloClasse in [LogisticRegression, DecisionTreeClassifier]:
    modelo = ModeloClasse()
    modelo.fit(X, y)
    previsao = modelo.predict([[3.5]])
    print(f"{ModeloClasse.__name__}: previsão para 3.5 = {previsao}")
````

**Contexto:** a PyData Streaming quer prever quais assinantes têm risco de cancelar (churn) com base
em horas assistidas e número de reclamações de suporte.

**Tarefas:**
1. Gerar um dataset sintético de assinantes.
2. Preparar os dados (normalização) e separar treino/teste.
3. Treinar um `DecisionTreeClassifier` para prever `cancelou`.
4. Avaliar o modelo com accuracy, precision e recall.
5. Verificar sinais de overfitting comparando treino vs. teste.

### ✅ Gabarito do projeto

````python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score

# 1. Dataset sintético
np.random.seed(42)
n = 300
horas_assistidas = np.random.normal(20, 8, n)
reclamacoes = np.random.poisson(0.8, n)
# Regra "real" (com ruído) para gerar o rótulo: poucas horas + muitas reclamações = maior risco de cancelar
probabilidade_cancelar = 1 / (1 + np.exp(-(reclamacoes * 1.2 - horas_assistidas * 0.08)))
cancelou = (probabilidade_cancelar > np.random.rand(n)).astype(int)

X = np.column_stack([horas_assistidas, reclamacoes])
y = cancelou

# 2. Preparação: normalização e split treino/teste
scaler = StandardScaler()
X_normalizado = scaler.fit_transform(X)
X_treino, X_teste, y_treino, y_teste = train_test_split(X_normalizado, y, test_size=0.25, random_state=42)

# 3. Treinamento
modelo = DecisionTreeClassifier(max_depth=4, random_state=42)
modelo.fit(X_treino, y_treino)

# 4. Avaliação
pred_treino = modelo.predict(X_treino)
pred_teste = modelo.predict(X_teste)

print("Avaliação no TREINO:")
print(f"  Accuracy: {accuracy_score(y_treino, pred_treino):.3f}")
print(f"  Precision: {precision_score(y_treino, pred_treino, zero_division=0):.3f}")
print(f"  Recall: {recall_score(y_treino, pred_treino, zero_division=0):.3f}")

print("\nAvaliação no TESTE:")
print(f"  Accuracy: {accuracy_score(y_teste, pred_teste):.3f}")
print(f"  Precision: {precision_score(y_teste, pred_teste, zero_division=0):.3f}")
print(f"  Recall: {recall_score(y_teste, pred_teste, zero_division=0):.3f}")

# 5. Verificando overfitting
diferenca_acuracia = accuracy_score(y_treino, pred_treino) - accuracy_score(y_teste, pred_teste)
print(f"\nDiferença treino-teste (accuracy): {diferenca_acuracia:.3f}")
if diferenca_acuracia > 0.15:
    print("Possível overfitting — considere reduzir max_depth.")
else:
    print("Sem sinais fortes de overfitting.")
````
