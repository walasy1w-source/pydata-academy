## Capítulo 1: Regressão Linear

A **regressão linear** modela a relação entre variáveis como uma linha reta:
`y = a*x + b`. O algoritmo encontra os melhores valores de `a` (coeficiente/inclinação) e `b`
(intercepto) que minimizam o erro entre a linha e os pontos reais. É o algoritmo mais simples e
interpretável para prever valores numéricos contínuos (preços, temperaturas, vendas).

**💻 Exemplo comentado:**

````python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1], [2], [3], [4], [5]])   # horas estudadas
y = np.array([50, 55, 65, 70, 80])         # nota obtida

modelo = LinearRegression()
modelo.fit(X, y)

print(f"Coeficiente (a): {modelo.coef_[0]:.2f}")
print(f"Intercepto (b): {modelo.intercept_:.2f}")
print(f"Previsão para 6 horas de estudo: {modelo.predict([[6]])[0]:.2f}")

from sklearn.metrics import r2_score
print(f"R² no próprio treino: {r2_score(y, modelo.predict(X)):.3f}")
````

## Capítulo 2: Regressão Logística

Apesar do nome, a **regressão logística** é usada para **classificação** (não
regressão de valores contínuos). Ela estima a probabilidade de um exemplo pertencer a uma classe,
usando uma função sigmoide que transforma qualquer valor em um número entre 0 e 1. Se a probabilidade
for maior que 0.5 (limiar padrão), o modelo prevê a classe positiva.

**💻 Exemplo comentado:**

````python
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])  # aprovado (1) ou reprovado (0), por exemplo

modelo = LogisticRegression()
modelo.fit(X, y)

print("Previsão para X=4.5:", modelo.predict([[4.5]]))
print("Probabilidades para X=4.5:", modelo.predict_proba([[4.5]]))

from sklearn.metrics import accuracy_score
print(f"Acurácia no treino: {accuracy_score(y, modelo.predict(X)):.2f}")
````

## Capítulo 3: Árvores de Decisão

Uma **árvore de decisão** aprende uma sequência de perguntas sim/não (ex.:
"reclamações > 2?") que dividem os dados progressivamente até chegar a uma previsão. É fácil de
interpretar (pode ser visualizada como um fluxograma) e funciona tanto para classificação
(`DecisionTreeClassifier`) quanto para regressão (`DecisionTreeRegressor`). O parâmetro `max_depth`
controla a profundidade da árvore — árvores muito profundas tendem a overfitting (Sprint 8).

**💻 Exemplo comentado:**

````python
from sklearn.tree import DecisionTreeClassifier
import numpy as np

X = np.array([[0], [1], [2], [3], [4], [5], [6], [7]])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

modelo = DecisionTreeClassifier(max_depth=2, random_state=0)
modelo.fit(X, y)

print("Previsão para X=3.5:", modelo.predict([[3.5]]))
print("Importância das features:", modelo.feature_importances_)

# A profundidade real da árvore treinada
print("Profundidade da árvore:", modelo.get_depth())
````

## Capítulo 4: Florestas Aleatórias (Random Forest)

Uma **Random Forest** treina **várias árvores de decisão diferentes** (cada uma
vendo uma amostra aleatória dos dados e das features) e combina as previsões (votação para
classificação, média para regressão). Essa combinação ("ensemble") costuma ser mais robusta e menos
propensa a overfitting do que uma única árvore, sendo um dos algoritmos mais usados para dados
tabulares no mercado.

**💻 Exemplo comentado:**

````python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

np.random.seed(0)
X = np.random.rand(100, 2)
y = (X[:, 0] + X[:, 1] > 1).astype(int)

modelo = RandomForestClassifier(n_estimators=50, random_state=0)
modelo.fit(X, y)

print("Número de árvores:", len(modelo.estimators_))
print("Importância das features:", modelo.feature_importances_)
print("Previsão para [[0.6, 0.6]]:", modelo.predict([[0.6, 0.6]]))
````

## Capítulo 5: Máquinas de Vetores de Suporte (SVM)

O **SVM (Support Vector Machine)** busca a "fronteira de decisão" que melhor separa
as classes, maximizando a margem (distância) entre a fronteira e os pontos mais próximos de cada
classe (os "vetores de suporte"). Com o parâmetro `kernel` (ex.: `'linear'`, `'rbf'`), o SVM também
consegue separar classes que não são linearmente separáveis, projetando os dados em um espaço de
dimensão maior.

**💻 Exemplo comentado:**

````python
from sklearn.svm import SVC
import numpy as np

X = np.array([[1, 1], [2, 1], [1, 2], [8, 8], [9, 8], [8, 9]])
y = np.array([0, 0, 0, 1, 1, 1])

modelo = SVC(kernel='linear')
modelo.fit(X, y)

print("Previsão para [[5, 5]]:", modelo.predict([[5, 5]]))
print("Previsão para [[1.5, 1.5]]:", modelo.predict([[1.5, 1.5]]))
print("Número de vetores de suporte:", modelo.n_support_)
````

## Capítulo 6: Validação Cruzada

Um único `train_test_split` pode dar uma avaliação enganosa, dependendo de qual
"corte" de dados caiu no teste (sorte ou azar). A **validação cruzada** (k-fold) resolve isso:
divide os dados em `k` partes (folds), treina o modelo `k` vezes, cada vez usando uma parte diferente
como teste e o resto como treino, e calcula a média (e desvio padrão) das métricas — uma avaliação
muito mais robusta e confiável do que um único split.

**💻 Exemplo comentado:**

````python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np

np.random.seed(0)
X = np.random.rand(100, 2)
y = (X[:, 0] + X[:, 1] > 1).astype(int)

modelo = DecisionTreeClassifier(max_depth=3, random_state=0)
scores = cross_val_score(modelo, X, y, cv=5)  # 5-fold cross-validation

print("Acurácia de cada fold:", scores.round(3))
print(f"Média: {scores.mean():.3f}, Desvio padrão: {scores.std():.3f}")
````

## Capítulo 7: Ajuste de Hiperparâmetros

**Hiperparâmetros** são configurações do algoritmo escolhidas *antes* do
treinamento (ex.: `max_depth` de uma árvore, `n_estimators` de uma floresta) — diferente dos
parâmetros que o modelo aprende sozinho (como os coeficientes da regressão linear). O
`GridSearchCV` testa sistematicamente várias combinações de hiperparâmetros, usando validação cruzada
em cada uma, e retorna a melhor combinação encontrada (`.best_params_`) e o melhor modelo
(`.best_estimator_`).

**💻 Exemplo comentado:**

````python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
import numpy as np

np.random.seed(0)
X = np.random.rand(100, 2)
y = (X[:, 0] + X[:, 1] > 1).astype(int)

grade_parametros = {"max_depth": [1, 2, 3, 5, None], "min_samples_split": [2, 5, 10]}

busca = GridSearchCV(DecisionTreeClassifier(random_state=0), grade_parametros, cv=5)
busca.fit(X, y)

print("Melhores parâmetros:", busca.best_params_)
print(f"Melhor acurácia média (validação cruzada): {busca.best_score_:.3f}")
````

````python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV

# 1. Dataset sintético
np.random.seed(42)
n = 200
horas_assistidas = np.random.normal(20, 8, n)
reclamacoes = np.random.poisson(0.8, n)
probabilidade_cancelar = 1 / (1 + np.exp(-(reclamacoes * 1.2 - horas_assistidas * 0.08)))
cancelou = (probabilidade_cancelar > np.random.rand(n)).astype(int)
X = np.column_stack([horas_assistidas, reclamacoes])
y = cancelou

# 2 e 3. Treinar e avaliar 3 modelos com validação cruzada
modelos = {
    "Regressão Logística": LogisticRegression(),
    "Árvore de Decisão": DecisionTreeClassifier(max_depth=4, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=50, max_depth=4, random_state=42),
}

resultados = {}
for nome, modelo in modelos.items():
    scores = cross_val_score(modelo, X, y, cv=5)
    resultados[nome] = scores.mean()
    print(f"{nome}: acurácia média = {scores.mean():.3f} (+/- {scores.std():.3f})")

# 4. Escolhendo o melhor modelo
melhor_modelo_nome = max(resultados, key=resultados.get)
print(f"\nMelhor modelo: {melhor_modelo_nome} (acurácia média: {resultados[melhor_modelo_nome]:.3f})")

# 5. Ajuste de hiperparâmetros do modelo vencedor (assumindo que seja baseado em árvore)
grade_parametros = {"max_depth": [2, 3, 4, 5], "n_estimators": [20, 50, 100]}
busca = GridSearchCV(RandomForestClassifier(random_state=42), grade_parametros, cv=5)
busca.fit(X, y)
print(f"\nMelhores hiperparâmetros (Random Forest): {busca.best_params_}")
print(f"Melhor acurácia média após ajuste: {busca.best_score_:.3f}")
````
