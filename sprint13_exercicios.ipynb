## Capítulo 1: Fundamentos de Séries Temporais

Uma **série temporal** é uma sequência de valores medidos ao longo do tempo, onde
**a ordem importa** — embaralhar os dados destruiria a informação (diferente da maioria dos datasets
vistos até aqui). Em pandas, usamos um índice de datas (`DatetimeIndex`, via `pd.date_range`).
Conceitos-chave: **defasagem/lag** (`.shift()`, o valor de N períodos atrás) e **autocorrelação**
(`.autocorr()`, o quanto o valor atual se correlaciona com valores passados — a base de quase toda
análise de séries temporais).

**💻 Exemplo comentado:**

````python
import pandas as pd
import numpy as np

datas = pd.date_range(start="2025-01-01", periods=12, freq="MS")   # início de cada mês
vendas = pd.Series([100, 110, 105, 120, 130, 125, 140, 150, 145, 160, 170, 165], index=datas)

print(vendas)
print("\nValor defasado em 1 período (lag 1):")
print(vendas.shift(1))
print(f"\nAutocorrelação (lag 1): {vendas.autocorr(lag=1):.3f}")
````

## Capítulo 2: Decomposição de Séries Temporais

Uma série temporal pode ser decomposta em três componentes: **tendência** (a
direção geral de longo prazo, subindo ou descendo), **sazonalidade** (padrões que se repetem em
intervalos fixos, como picos de vendas todo dezembro) e **ruído/resíduo** (a variação aleatória que
sobra depois de remover tendência e sazonalidade). `statsmodels.tsa.seasonal.seasonal_decompose`
calcula essa decomposição automaticamente.

**💻 Exemplo comentado:**

````python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose

np.random.seed(0)
datas = pd.date_range(start="2022-01-01", periods=48, freq="MS")
tendencia = np.linspace(100, 200, 48)
sazonalidade = 10 * np.sin(np.arange(48) * (2 * np.pi / 12))
ruido = np.random.normal(0, 3, 48)
serie = pd.Series(tendencia + sazonalidade + ruido, index=datas)

decomposicao = seasonal_decompose(serie, model="additive", period=12)

print("Tendência (primeiros valores):")
print(decomposicao.trend.dropna().head(3))
print("\nSazonalidade (primeiros valores):")
print(decomposicao.seasonal.head(3))
````

## Capítulo 3: Modelos ARIMA

**ARIMA** (AutoRegressive Integrated Moving Average) combina três ideias: **AR**
(autoregressivo — o valor atual depende de valores passados da própria série), **I** (integrado —
aplicamos diferenciação, `.diff()`, para tornar a série estacionária, removendo tendência), e **MA**
(médias móveis — o valor atual depende de erros de previsão passados). A notação `ARIMA(p, d, q)`
define a ordem de cada componente. Em `statsmodels`, ajustamos com `ARIMA(serie, order=(p,d,q)).fit()`
e prevemos com `.forecast(passos)`.

**💻 Exemplo comentado:**

````python
import pandas as pd
import numpy as np
import warnings
from statsmodels.tsa.arima.model import ARIMA

warnings.filterwarnings("ignore")   # statsmodels emite avisos de convergência, ignoramos por simplicidade

np.random.seed(0)
datas = pd.date_range(start="2023-01-01", periods=36, freq="MS")
serie = pd.Series(np.linspace(100, 150, 36) + np.random.normal(0, 3, 36), index=datas)

modelo = ARIMA(serie, order=(1, 1, 1))
modelo_ajustado = modelo.fit()

previsao = modelo_ajustado.forecast(steps=3)
print("Previsão para os próximos 3 meses:")
print(previsao)
````

## Capítulo 4: Previsão com Séries Temporais

Avaliar um modelo de série temporal exige cuidado: **nunca embaralhamos** os dados
antes de dividir treino/teste (Sprint 8) — o teste deve ser sempre o período **mais recente**, e o
treino, o passado. Avaliamos a previsão com as mesmas métricas de regressão (MAE, RMSE — Sprint 4/8),
comparando os valores previstos com os valores reais do período de teste.

**💻 Exemplo comentado:**

````python
import pandas as pd
import numpy as np
import warnings
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

warnings.filterwarnings("ignore")

np.random.seed(0)
datas = pd.date_range(start="2022-01-01", periods=36, freq="MS")
serie = pd.Series(np.linspace(100, 200, 36) + np.random.normal(0, 5, 36), index=datas)

# Split respeitando a ordem: treino = passado, teste = período mais recente
tamanho_teste = 6
treino = serie.iloc[:-tamanho_teste]
teste = serie.iloc[-tamanho_teste:]

modelo = ARIMA(treino, order=(1, 1, 1)).fit()
previsao = modelo.forecast(steps=tamanho_teste)

mae = mean_absolute_error(teste, previsao)
print(f"MAE da previsão: {mae:.2f}")
````

## 🏆 Capítulo 5 — Projeto: Previsão de Demanda

**Contexto:** a PyData Store quer prever a demanda de um produto para os próximos 3 meses, usando o
histórico de vendas mensais.

**Tarefas:**
1. Gerar um histórico sintético de 36 meses (tendência + sazonalidade + ruído).
2. Decompor a série para confirmar a presença de tendência e sazonalidade.
3. Dividir em treino/teste (últimos 6 meses como teste) e ajustar um modelo ARIMA.
4. Avaliar o modelo no teste (MAE) e gerar a previsão final para os próximos 3 meses.

````python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

# 1. Histórico sintético de vendas (36 meses)
np.random.seed(42)
datas = pd.date_range(start="2022-01-01", periods=36, freq="MS")
tendencia = np.linspace(1000, 2200, 36)
sazonalidade = 150 * np.sin(np.arange(36) * (2 * np.pi / 12))
ruido = np.random.normal(0, 30, 36)
vendas = pd.Series(tendencia + sazonalidade + ruido, index=datas)
print("1. Série de vendas gerada com", len(vendas), "meses de histórico.")

# 2. Decomposição
decomposicao = seasonal_decompose(vendas, model="additive", period=12)
amplitude_tendencia = decomposicao.trend.dropna().max() - decomposicao.trend.dropna().min()
amplitude_sazonal = decomposicao.seasonal.max() - decomposicao.seasonal.min()
print(f"2. Amplitude da tendência: {amplitude_tendencia:.2f}, Amplitude sazonal: {amplitude_sazonal:.2f}")

# 3. Split treino/teste e ajuste do ARIMA
tamanho_teste = 6
treino = vendas.iloc[:-tamanho_teste]
teste = vendas.iloc[-tamanho_teste:]

modelo = ARIMA(treino, order=(1, 1, 1)).fit()
previsao_teste = modelo.forecast(steps=tamanho_teste)

# 4. Avaliação e previsão final
mae = mean_absolute_error(teste, previsao_teste)
print(f"3 e 4. MAE no conjunto de teste: {mae:.2f}")

# Reajustando com todos os dados disponíveis para prever o futuro real
modelo_final = ARIMA(vendas, order=(1, 1, 1)).fit()
previsao_futura = modelo_final.forecast(steps=3)
print("\nPrevisão para os próximos 3 meses:")
print(previsao_futura.round(2))
````
