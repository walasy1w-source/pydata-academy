## Capítulo 2: Variáveis e Tipos de Dados

Uma variável é como uma caixa com etiqueta: você guarda um valor dentro dela e usa o
nome da etiqueta para acessá-lo depois. Cada valor tem um **tipo**: números inteiros são `int`,
números com casas decimais são `float`, textos são `str` e valores verdadeiro/falso são `bool`.

📌 A função `type()` revela o tipo de qualquer valor. Às vezes precisamos **converter** um tipo em
outro (*casting*) — por exemplo, transformar o texto `"25"` no número `25` com `int()`.

**💻 Exemplo comentado:**

````python
idade = 25           # int - número inteiro
altura = 1.68        # float - número decimal
nome = "Carla"       # str - texto
estudante = True     # bool - verdadeiro/falso

print(type(idade), type(altura), type(nome), type(estudante))

# Convertendo tipos (casting)
idade_texto = str(idade)
nota_texto = "9.5"
nota_numero = float(nota_texto)
print("Nota como número:", nota_numero)
````

## Capítulo 3: Operadores

Operadores são as ferramentas da caixa de ferramentas do Python. Os **aritméticos**
(`+ - * / // % **`) fazem contas. Os **de comparação** (`== != > < >= <=`) comparam dois valores e
retornam `True` ou `False`. Os **lógicos** (`and`, `or`, `not`) combinam condições.

📌 Existem também dois operadores especiais: `in` (verifica se um valor está dentro de uma coleção,
como uma lista) e `is` (verifica se duas variáveis apontam para o **mesmo objeto** na memória — muito
usado para comparar com `None`).

**💻 Exemplo comentado:**

````python
a = 10
b = 3

print(a + b, a - b, a * b, a / b)   # aritméticos básicos
print(a // b)    # divisão inteira -> 3
print(a % b)     # resto da divisão (módulo) -> 1
print(a ** b)    # potência -> 1000

print(a > b, a == b, a != b)   # comparação

x = True
y = False
print(x and y, x or y, not x)  # lógicos
````

## Capítulo 5: Estruturas Condicionais

Estruturas condicionais são como um semáforo: dependendo de uma condição, o programa
segue por um caminho ou por outro. O `if` testa a primeira condição; se for falsa, o `elif` testa
outra; e o `else` captura tudo que sobrou.

📌 Condições podem ser **aninhadas** (um `if` dentro de outro) ou encadeadas em sequência com vários
`elif`, o que é útil para classificar algo em mais de duas categorias (ex.: faixas etárias).

**💻 Exemplo comentado:**

````python
nota = 7.5

if nota >= 9:
    conceito = "Excelente"
elif nota >= 7:
    conceito = "Bom"
elif nota >= 5:
    conceito = "Regular"
else:
    conceito = "Precisa melhorar"

print(f"Seu conceito é: {conceito}")
````

## Capítulo 6: Estruturas de Repetição

Loops servem para repetir uma ação várias vezes. O `for` é ideal quando sabemos
quantas vezes repetir (ou percorremos uma sequência). O `while` repete enquanto uma condição for
verdadeira — útil quando não sabemos o número exato de repetições, como em um sistema de tentativas
de login.

⚠️ Cuidado com o **loop infinito**: se a condição do `while` nunca deixar de ser verdadeira (por
exemplo, porque você esqueceu de atualizar uma variável dentro do loop), o programa nunca para.

**💻 Exemplo comentado:**

````python
# for com range
for i in range(1, 6):
    print(f"Repetição número {i}")

# while
contador = 0
while contador < 3:
    print("Contador:", contador)
    contador += 1

# break e continue
for numero in range(10):
    if numero == 5:
        break        # interrompe o loop
    if numero % 2 == 0:
        continue     # pula para a próxima iteração
    print(numero)
````

## Capítulo 7: Strings e Manipulação de Texto

Uma string pode ser vista como um colar de contas, onde cada "conta" é um caractere
com uma posição (índice), começando em 0. Podemos usar métodos prontos como `.upper()`, `.lower()`,
`.strip()` (remove espaços das pontas), `.split()` (separa em lista) e `.replace()` (substitui
trechos) para limpar e transformar textos — algo muito comum ao tratar dados reais, que quase sempre
vêm "sujos" (com espaços extras, maiúsculas/minúsculas inconsistentes etc).

**💻 Exemplo comentado:**

````python
frase = "  Python é incrível "

print(frase.strip())          # remove espaços das pontas
print(frase.upper())          # tudo maiúsculo
print(frase.lower())          # tudo minúsculo
print(frase.strip().split())  # separa em lista de palavras

nome = "Maria Silva"
print(nome[0:5])              # fatiamento: "Maria"
print(nome.replace("Silva", "Souza"))
print("-".join(["2026", "07", "02"]))
````

## Capítulo 8: Listas, Tuplas e Dicionários

Uma **lista** é como uma lista de compras: ordenada e editável. Uma **tupla** é
parecida, mas imutável. Um **dicionário** guarda pares chave-valor, como uma agenda de contatos.

📌 Uma **lista de dicionários** (como `[{'produto': 'A', 'quantidade': 10}, ...]`) é a estrutura mais
comum para representar dados tabulares em Python puro — é assim que pandas representa um DataFrame
por baixo dos panos. A função `zip()` combina duas listas elemento a elemento, muito útil para criar
um dicionário a partir de duas listas paralelas.

**💻 Exemplo comentado:**

````python
# Lista: coleção editável e ordenada
frutas = ["maçã", "banana", "uva"]
frutas.append("laranja")

# Tupla: coleção ordenada e IMUTÁVEL
coordenadas = (10.5, -23.4)

# Dicionário: pares chave-valor
aluno = {
    "nome": "Beatriz",
    "idade": 22,
    "curso": "Ciência de Dados"
}

print(frutas)
print(coordenadas)
print(aluno["nome"], "-", aluno["curso"])

aluno["idade"] = 23           # atualiza um valor
aluno["cidade"] = "Recife"    # adiciona uma nova chave
print(aluno)
````

## Capítulo 9: Funções

Uma função é como uma máquina de suco: você coloca frutas (os parâmetros) e recebe um
suco pronto (o retorno). Uma função pode retornar **múltiplos valores organizados**, por exemplo em
um dicionário, o que é muito útil para resumir dados (total, média, máximo, mínimo).

📌 Funções **lambda** são funções pequenas, de uma linha, sem nome próprio — muito usadas junto com
`map()` (aplica uma função a cada item de uma lista) e `filter()` (mantém apenas os itens que
atendem a uma condição). Isso é uma primeira introdução à **programação funcional**.

**💻 Exemplo comentado:**

````python
def saudacao(nome, saudacao_texto="Olá"):
    """Retorna uma saudação personalizada."""
    return f"{saudacao_texto}, {nome}!"

print(saudacao("Pedro"))
print(saudacao("Ana", "Bom dia"))

def calcular_media(notas):
    return sum(notas) / len(notas)

minhas_notas = [8, 7.5, 9, 6]
print(f"Média: {calcular_media(minhas_notas):.2f}")
````

### ✅ Capítulo 8: Listas, Tuplas e Dicionários — Soluções

````python
# --- Exercício 1 ---
cores = ("azul", "dourado", "branco")
# cores[0] = "verde"  # gera TypeError: tuplas são imutáveis

# --- Exercício 2 ---
vendas = [
    {"produto": "A", "quantidade": 10, "valor_unitario": 5.0},
    {"produto": "B", "quantidade": 3, "valor_unitario": 20.0},
    {"produto": "A", "quantidade": 7, "valor_unitario": 5.0},
]

valor_total = sum(v["quantidade"] * v["valor_unitario"] for v in vendas)
print(f"Valor total vendido: R$ {valor_total:.2f}")

quantidade_por_produto = {}
for v in vendas:
    quantidade_por_produto[v["produto"]] = quantidade_por_produto.get(v["produto"], 0) + v["quantidade"]
mais_vendido = max(quantidade_por_produto, key=quantidade_por_produto.get)
print(f"Produto mais vendido: {mais_vendido} ({quantidade_por_produto[mais_vendido]} unidades)")

# --- Exercício 3 ---
nomes = ["Ana", "Bruno", "Carla"]
idades = [23, 31, 27]
pessoas = dict(zip(nomes, idades))
print(pessoas)

# --- Exercício 4 ---
livro = {"titulo": "Python Fluente", "autor": "Luciano Ramalho", "ano": 2015}
print("autor" in livro)

# --- Exercício 5 ---
livro = {"titulo": "Python Fluente", "autor": "Luciano Ramalho", "ano": 2015}
for chave, valor in livro.items():
    print(chave, ":", valor)

# --- Exercício 6 ---
livro = {"titulo": "Python Fluente"}
print(livro.get("autor", "Autor não informado"))
````

### ✅ Capítulo 9: Funções — Soluções

````python
# --- Exercício 1 ---
def somar(a, b):
    return a + b

print(somar(3, 4))

# --- Exercício 2 ---
def resumo_vendas(vendas):
    return {
        "total": sum(vendas),
        "media": sum(vendas) / len(vendas),
        "maximo": max(vendas),
        "minimo": min(vendas),
    }

vendas = [120, 340, 95, 210, 180]
print(resumo_vendas(vendas))

# --- Exercício 3 ---
numeros = [1, 2, 3, 4, 5]
quadrados = list(map(lambda n: n ** 2, numeros))
pares = list(filter(lambda n: n % 2 == 0, numeros))
print("Quadrados:", quadrados)
print("Pares:", pares)

# --- Exercício 4 ---
def formatar_preco(valor, moeda="R$"):
    return f"{moeda} {valor:.2f}"

print(formatar_preco(49.9))

# --- Exercício 5 ---
def contar_vogais(palavra):
    vogais = "aeiouAEIOU"
    return sum(1 for letra in palavra if letra in vogais)

print(contar_vogais("Python"))

# --- Exercício 6 ---
def dobro(n):
    resultado = n * 2
    return resultado

print(dobro(4))
````
